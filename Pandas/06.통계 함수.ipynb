```python
# 데이터 불러오기 & 확인
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

file_path = '/content/drive/MyDrive/github/딥러닝을 위한 파이썬/Pandas/data/05000266/part3/auto-mpg.csv'
df = pd.read_csv(file_path)

df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'name']
df.head()
```

```python
# 각 열의 평균
print(f'df.mean:\n{df.mean()}\n')

# 특정 열의 평균
print(f'df["mpg"].mean:\n{df["mpg"].mean()}')
```

```python
# 각 열의 중간값
print(f'df.median:\n{df.median()}\n')

# 특정 열의 중간값
print(f'df[["mpg", "weight"]].median:\n{df[["mpg", "weight"]].median()}')
```

```python
# 각 열의 최댓값
print(f'df.max:\n{df.max()}\n')

# 특정 열의 최댓값
print(f'df[["mpg"]].max:\n{df["mpg"].max()}')
```

```python
# 각 열의 최솟값
print(f'df.min:\n{df.min()}\n')

# 특정 열의 최솟값
print(f'df[["mpg"]].min:\n{df["mpg"].min()}')
```

```python
# 각 열의 표준편차
print(f'df.std:\n{df.std()}\n')

# 특정 열의 표준편차
print(f'df[["mpg"]].std:\n{df["mpg"].std()}')
```

```python
# 모든 열의 상관계수
print(f'df.corr:\n{df.corr()}\n')

# 특정 두 열의 상관계수
print(f'df[["mpg", "weight"]].corr:\n{df[["mpg", "weight"]].corr()}')
```
