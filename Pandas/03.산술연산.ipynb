## 1.시리즈 연산

### 1-1.시리즈 vs 숫자


```python
import pandas as pd

# Dictionary -> Series
student = pd.Series({'국어':75, '수학':80, '영어':78})
print(f'DataFrame:\n{student}\n')

add_20 = student + 20
print(f'student + 20:\n{add_20}\n')

sub_20 = student - 20
print(f'student - 20:\n{sub_20}\n')

mul_2 = student*2
print(f'student*2:\n{mul_2}\n')

div_2 = student/2
print(f'student/2:\n{div_2}')
```

### 1-2.시리즈 vs 시리즈

```python
import numpy as np

student1 = pd.Series({'국어':75, '수학':80, '영어':78})
student2 = pd.Series({'영어':75, '국어':80, '수학':78})
print(f'student1:\n{student1}\n')
print(f'student2:\n{student2}\n')

add = student1 + student2
sub = student1 - student2
mul = student1 * student2
div = student1 / student2

result = pd.DataFrame([add, sub, mul, div], index = ['덧셈', '뺄셈', '곱셈', '나눗셈'])
print(f'사칙연산 결과:\n{result}')
```

```python
# 인덱스가 대응되지 않거나 NaN 값이 포함된 경우
import pandas as pd
import numpy as np

sr1 = pd.Series({'A':10, 'B':20, 'C':30, 'D':np.nan})
sr2 = pd.Series({ 'B':20, 'C':30, 'D':30, 'E':40})
print(f'sr1:\n{sr1}\n')
print(f'sr2:\n{sr2}\n')

add = sr1 + sr2
sub = sr1 - sr2
mul = sr1 * sr2
div = sr1 / sr2

result = pd.DataFrame([add, sub, mul, div], index=['덧셈', '뺄셈', '곱셈', '나눗셈'])
print(f'사칙연산 결과:\n{result}')
```

### 1-3.연산 메소드


```python
import pandas as pd
import numpy as np

sr1 = pd.Series({'A':10, 'B':20, 'C':30, 'D':np.nan})
sr2 = pd.Series({ 'B':20, 'C':30, 'D':30, 'E':40})
print(f'sr1:\n{sr1}\n')
print(f'sr2:\n{sr2}\n')

add = sr1.add(sr2, fill_value=0) # fill_value: 인덱스가 대응되지 않거나 NaN 값이 포함된 경우 NaN 값이 리턴 되는것을 방지
sub = sr1.sub(sr2, fill_value=0)
mul = sr1.mul(sr2, fill_value=0)
div = sr1.div(sr2, fill_value=0)

result = pd.DataFrame([add, sub, mul, div], index=['덧셈', '뺄셈', '곱셈', '나눗셈'])
print(f'사칙연산 결과:\n{result}')
```

## 2.데이터프레임 연산

### 2-1.데이터프레임 vs 숫자

```python
import numpy as np
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','fare']]
print(f'DataFrame:\n{df.head()}\n')

add_10 = df + 10
print(f'df + 10:\n{add_10.head()}\n')

sub_10 = df - 10
print(f'df - 10:\n{sub_10.head()}\n')

mul_10 = df * 10
print(f'df * 10:\n{mul_10.head()}\n')

div_10 = df / 10
print(f'df / 10:\n{div_10.head()}\n')
```

### 2-2.데이터프레임 vs 데이터프레임

```python
import numpy as np
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age','fare']]
print(f'DataFrame:\n{df.tail()}\n')

add_10 = df + 10
print(f'df + 10:\n{add_10.tail()}\n')

add = df + add_10
print(f'df + add_10:\n{add.tail()}')
```
