### 2-2.웹 스크래핑

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# Scraping(데이터 가져오기)
url = 'http://www.kric.go.kr/jsp/industry/rss/citystapassList.jsp?q_org_cd=A010010021&q_fdate=2020'
html_text = requests.get(url).text

# Parsing(데이터 구분)
soup = BeautifulSoup(html_text, 'html.parser')

# Manipulation(데이터 가공)
table = soup.find('table', {'class':'listtbl_c100'})
trs = table.find('tbody').find_all('tr')    # tbody안에 모든 tr을 가져옴

stationpassengers = []
for tr in trs[1:]:
    dic = {}
    tds = tr.find_all('td')
    dic['station'] = tds[0].text
    dic['승차인원'] = tds[2].text
    dic['하차인원'] = tds[3].text
    stationpassengers.append(dic)

df = pd.DataFrame(stationpassengers)
df = df.set_index('station')
print(f'DataFrame:\n{df}')
```

## 3.Open API 활용하여 데이터 수집하기
[실시간 영업소간 통행시간](http://data.ex.co.kr/openapi/trtm/realUnitTrtm)

```python
import requests
import pandas as pd

key = 'my key' 
StartUnitCode = '101' # 서울
EndUnitCode = '140'   # 부산

# 데이터 가져오기
URL = 'http://data.ex.co.kr/openapi/trtm/realUnitTrtm?key='
url = URL + key + '&type=json&iStartUnitCode=' + StartUnitCode + '&iEndUnitCode=' + EndUnitCode

response = requests.get(url)
json = response.json()
cars = json['realUnitTrtmVO']

# 리스트에 데이터 넣기
records = []
for car in cars:
    dic = {}
    dic['Date'] = car['stdDate']
    dic['Time'] = car['stdTime']
    dic['CarType'] = car['tcsCarTypeDivName']
    records.append(dic)

# DataFrame으로 만들기
df = pd.DataFrame(records)
print(f'서울 -> 부산 실시간 영업소간 통행시간:\n{df}')
```

### 4-4.여러개의 데이터 프레임을 하나의 Excel 파일로 저장

```python
import pandas as pd

data1 = {'이름': ['최성욱', '박솔아', '이상정'],
        '국어': [80, 90, 85],
        '수학': [100, 82, 93],
        '영어': [70, 93, 75]}

data2 = {'c0':[1,2,3],
         'c1':[4,5,6],
         'c2':[7,8,9],
         'c3':[10,11,12],
         'c4':[13,14,15],}

df1 = pd.DataFrame(data1).set_index('이름')
df2 = pd.DataFrame(data2)
print(f'DataFrame1:\n{df1}\n')
print(f'DataFrame2:\n{df2}')

# Excelwriter 생성후 데이터프레임 Excelwriter에 저장
writer = pd.ExcelWriter('./excelwriter.xlsx')
df1.to_excel(writer, sheet_name='sheet1')
df2.to_excel(writer, sheet_name='sheet2')
writer.save()
```
