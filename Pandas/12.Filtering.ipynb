## 1.불린 인덱싱

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')

# 10 <= age < 20
mask1 = (titanic['age'] >= 10) & (titanic['age'] < 20)
df_teenage = titanic.loc[mask1, :]
print(f"10<= age <20:\n{df_teenage.head()}\n")

# age < 10 and sex == female
mask2 = (titanic['age'] < 10) & (titanic['sex'] == 'female')
df_female_under10 = titanic.loc[mask2, :]
print(f"age < 10 and sex == female:\n{df_female_under10.head()}\n")

# age < 10 or age > 60
mask3 = (titanic['age'] < 10) | (titanic['age'] > 60)
df_under10_morethan60 = titanic.loc[mask3, :]
print(f"age < 10 or age > 60:\n{df_under10_morethan60.head()}")
```

## 2.isin() 메소드 활용

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')

# 불린 인덱싱
mask = (titanic['sibsp'] == 3) | (titanic['sibsp'] == 4) | (titanic['sibsp'] == 5)
df_bool = titanic.loc[mask, :]
print(f"boolean indexing:\n{df_bool.head()}\n")

# isin() 활용
isin_filter = titanic['sibsp'].isin([3, 4, 5])
df_isin = titanic.loc[isin_filter, :]
print(f"isin:\n{df_isin.head()}")
```
