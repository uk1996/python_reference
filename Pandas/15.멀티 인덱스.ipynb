```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'sex', 'class', 'fare', 'survived']]

grouped = df.groupby(['class', 'sex'])

# 그룹별 평균 구하기
g_mean = grouped.mean()
print(f"그룹별 평균:\n{g_mean}\n")

# class가 First인 행을 선택
print(f"class:First\n{g_mean.loc['First']}\n")

# class: First sex:male 선택
print(f"class:First, sex:male\n{g_mean.loc[('First', 'male')]}\n")

# xs 인덱서 사용
print(g_mean.xs('male', level='sex'))
```
