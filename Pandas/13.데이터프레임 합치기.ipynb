## 1.데이터프레임 연결

```python
import pandas as pd


df1 = pd.DataFrame({'a':[0, 1, 2, 3],
                    'b':[4, 5, 6, 7],
                    'c':[8, 9, 10, 11]})
print(f"df1:\n{df1}\n")

df2 = pd.DataFrame({'b':[0, 1, 2, 3, 4],
                    'c':[5, 6, 7, 8, 9],
                    'd':[10, 11, 12, 13, 14]},
                   index = [2, 3, 4, 5 ,6])
print(f"df2:\n{df2}\n")

print(f"pd.concat([df1, df2], axis=0):\n{pd.concat([df1, df2], axis=0)}\n")
print(f"pd.concat([df1, df2], axis=0, join='inner'):\n{pd.concat([df1, df2], axis=0, join='inner')}\n")
print(f"pd.concat([df1, df2], axis=0, join='inner', ignore_index=True):\n{pd.concat([df1, df2], axis=0, join='inner', ignore_index=True)}\n")

print(f"pd.concat([df1, df2], axis=1):\n{pd.concat([df1, df2], axis=1)}\n")
print(f"pd.concat([df1, df2], axis=1, join='inner'):\n{pd.concat([df1, df2], axis=1, join='inner')}")
```

```python
sr1 = pd.Series(['e0', 'e1', 'e2', 'e3'], name='e')
sr2 = pd.Series(['f0', 'f1', 'f2',], name='f')
sr3 = pd.Series(['g0', 'g1', 'g2', 'g3'], name='g')

print(f"pd.concat([df1, sr1], axis=1):\n{pd.concat([df1, sr1], axis=1)}\n")
print(f"pd.concat([sr1, sr2],axis=1):\n{pd.concat([sr1, sr2], axis=1)}\n")
print(f"pd.concat([sr1, sr2],axis=0):\n{pd.concat([sr1, sr2], axis=0)}")
```

## 2.데이터프레임 병합

```python
import pandas as pd
pd.set_option('display.unicode.east_asian_width', True) # 유니코드 사용 너비 조정

df1 = pd.read_excel('/content/drive/MyDrive/github/딥러닝을 위한 파이썬/Pandas/data/05000266/part6/stock price.xlsx')
df2 = pd.read_excel('/content/drive/MyDrive/github/딥러닝을 위한 파이썬/Pandas/data/05000266/part6/stock valuation.xlsx')

print(f"df1:\n{df1}\n")
print(f"df2:\n{df2}")
```

```python
pd.merge(df1, df2, how='inner', on=None)
```

```python
pd.merge(df1, df2, how='outer', on='id')
```

```python
pd.merge(df1, df2, how='left', left_on='stock_name', right_on='name')
```

```python
pd.merge(df1, df2, how='right', left_on='stock_name', right_on='name')
```

```python
# boolean indexing
price = df1.loc[df1['price'] < 50000]
print(f"price < 50000:\n{price}\n")

pd.merge(price, df2, how='inner', on='id')
```
