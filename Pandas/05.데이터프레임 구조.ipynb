## 2.데이터 요약 정보 확인


```python
import pandas as pd

# 데이터 불러오기
file_path = '/content/drive/MyDrive/github/딥러닝을 위한 파이썬/Pandas/data/05000266/part3/auto-mpg.csv'
df = pd.read_csv(file_path)

# 데이터프레임 컬럼명 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'name']

# 데이터프레임 크기(행, 열) 확인
print(f'df.shape: {df.shape}\n')

# 데이터 프레임 기본 정보
print({df.info()},'\n')

# 데이터 프레임 자료형 확인
print(f'df.dtypes:\n{df.dtypes}\n')

# 열의 자료형 확인
print(f'df["mpg"].dtypes: {df["mpg"].dtypes}\n')

# 데이터 프레임 기술 통계 정보 확인(산술 데이터)
print(f'df.describe:\n{df.describe()}\n')

# 데이터 프레임 기술 통계 정보 확인(열에 대한 정보)
print(f'df.describe(include="all"):\n{df.describe(include="all")}')
```

## 3.데이터 개수 확인

```python
import pandas as pd

# 데이터 불러오기
file_path = '/content/drive/MyDrive/github/딥러닝을 위한 파이썬/Pandas/data/05000266/part3/auto-mpg.csv'
df = pd.read_csv(file_path)

# 데이터프레임 컬럼명 지정
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'name']

# 데이터 프레임 각 열의 원소 개수
print(f'df.count:\n{df.count()}, type: {type(df.count())}\n')

# 열의 고유값(unique) 개수
print(f"df['cylinders'].value_counts():\n{df['cylinders'].value_counts()}")
```
