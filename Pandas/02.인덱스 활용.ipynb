## 1.특정 열을 행 인덱스로 설정
DataFrame.set_index(['열 이름'] or '열 이름')

```python
import pandas as pd

exam_data = {'이름': ['성욱', '솔아', '상정', '진영'],
             '수학': [90, 70, 75, 80],
             '영어': [70, 68, 90, 75],
             '국어': [80, 95, 70, 80],
             '과학': [95, 60, 86, 90]}

df = pd.DataFrame(exam_data)
print(f'DataFrame:\n{df}\n')

df = df.set_index('이름')
print(f'DataFrame:\n{df}\n')

df = df.set_index(['수학'])
print(f'DataFrame:\n{df}\n')

df = df.set_index(['영어', '국어'])
print(f'DataFrame:\n{df}')
```

## 2.행 인덱스 재배열
DataFrame.reindex(새로운 인덱스 배열)

```python
import pandas as pd

dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9], 'c3':[10,11,12], 'c4':[13,14,15]}
df = pd.DataFrame(dict_data, index = ['r0','r1','r2'])
print(f'DataFrame:\n{df}\n')

df2 = df.reindex(['r0','r1','r2','r3','r4'])
print(f'DataFrame:\n{df2}\n')

df3 = df.reindex(['r0','r1','r2','r3','r4'], fill_value=0)
print(f'DataFrame:\n{df3}')
```

## 3.행 인덱스 초기화
DataFrame.reset_index()

```python
import pandas as pd

dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9], 'c3':[10,11,12], 'c4':[13,14,15]}
df = pd.DataFrame(dict_data, index=['r0','r1','r2'])
print(f'DataFrame:\n{df}\n')

df = df.reset_index()
print(f'DataFrame:\n{df}')
```

## 4.행 인덕스를 기준으로 데이터프레임 정렬
DataFrame.sort_index(ascending = True or False)

```python
import pandas as pd

dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9], 'c3':[10,11,12], 'c4':[13,14,15]}
df = pd.DataFrame(dict_data, index=['r0','r1','r2'])
print(f'DataFrame:\n{df}\n')

df = df.sort_index(ascending=False) # (True: 오름차순, False: 내림차순)
print(f'DataFrame:\n{df}')
```

## 5.특정 열의 데이터 값을 기준으로 데이터프레임 정렬
DataFrame.sort_values(by = '열 이름', ascending = True or False)

```python
import pandas as pd

dict_data = {'c0':[1,2,3], 'c1':[4,5,6], 'c2':[7,8,9], 'c3':[10,11,12], 'c4':[13,14,15]}
df = pd.DataFrame(dict_data, index=['r0','r1','r2'])
print(f'DataFrame:\n{df}\n')

df = df.sort_values(by='c1', ascending=False)
print(f'DataFrame:\n{df}')
```
