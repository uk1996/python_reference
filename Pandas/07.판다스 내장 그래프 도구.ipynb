## 1.남북한 발전 전력량

```python
# 데이터 불러오기 & 확인
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd

file_path = '/content/drive/MyDrive/github/딥러닝을 위한 파이썬/Pandas/data/05000266/part3/남북한발전전력량.xlsx'
df = pd.read_excel(file_path)
df = df.iloc[[0, 5], 3:]
df.index = ['South', 'North']
df = df.T

df
```

## 2.자동차 연비


```python
# 데이터 불러오기 & 확인
import pandas as pd

file_path = '/content/drive/MyDrive/github/딥러닝을 위한 파이썬/Pandas/data/05000266/part3/auto-mpg.csv'
df = pd.read_csv(file_path)
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration', 'model_year', 'origin', 'name']
df
```

```python
# 산점도
df.plot(x = 'weight', y = 'mpg', kind = 'scatter') # 무게와 연비는 음의 상관계수를 가진다
```

```python
# 박스 플롯
df[['mpg', 'cylinders']].plot(kind = 'box')
```
