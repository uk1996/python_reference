## 1.열 순서 변경

```python
import pandas as pd
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[0:4, 'survived':'age']
print(f"dataFrame:\n{df}\n")

# column 리스트 만들기
columns = list(df.columns.values)
print(f"columns: {columns}\n")

# column 알파벳 순으로 정렬
columns_sorted = sorted(columns)
df_sorted = df[columns_sorted]
print(f"DataFrame_sorted:\n{df_sorted}\n")

# column 역순으로 정렬
columns_reversed =  list(reversed(columns))
df_reversed = df[columns_reversed]
print(f"DataFrame_reversed:\n{df_reversed}\n")

# column 임의의 순서로 재배치
columns_customed = ['sex', 'survived', 'age', 'pclass']
df_customed = df[columns_customed]
print(f"DataFrame_reversed:\n{df_customed}")
```

## 2. 열 분리

```python
import pandas as pd

df = pd.read_excel('/content/drive/MyDrive/github/딥러닝을 위한 파이썬/Pandas/data/05000266/part6/주가데이터.xlsx')
print(f"df_dtypes:\n{df.dtypes}\n")

# 연, 월, 일 분리
df['연월일'] = df['연월일'].astype('str') # dtype: datatime64 -> str
dates = df['연월일'].str.split('-') # 문자열 분리
print(f"dates:\n{dates.head()}, type: {type(dates)}\n")

# 데이터 프레임에 추가
df['연'] = dates.str.get(0)
df['월'] = dates.str.get(1)
df['일'] = dates.str.get(2)
df.head()
```
