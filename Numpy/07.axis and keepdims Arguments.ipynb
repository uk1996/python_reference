## 1.행렬 ndarray에서의 axis, keepdims

### 1-1.np.sum and ndarray.sum of Vector

```python
import numpy as np

a = np.arange(5)
print(f'ndarray: {a}\n')

print(f'np.sum: {np.sum(a)}')
print(f'ndarray.sum: {a.sum()}')
```

### 1-2.np.sum and ndarray.sum of Matrix

```python
import numpy as np

a = np.arange(6).reshape((2,-1))
print(f'ndarray:\n{a}\n')

print(f'np.sum: {np.sum(a)}')
print(f'ndarray.sum: {a.sum()}')
```

### 1-3.axis Argument

```python
import numpy as np

a = np.arange(12).reshape((3,-1))
print(f'scores:\n{a}\n')

sum_class = a.sum(axis=0)
print(f'sum_class(axis=0): {sum_class}, shape: {sum_class.shape}')

sum_student = a.sum(axis=1)
print(f'sum_student(axis=1): {sum_student}, shape: {sum_student.shape}')
```

### 1-4.Keepdims Argument

```python
import numpy as np

a = np.arange(12).reshape((3,-1))
print(f'scores:\n{a}\n')

sum_class = np.sum(a, axis=0, keepdims=True)
print(f'sum_class(axis=0):\n{sum_class}, shape: {sum_class.shape}')

sum_student = np.sum(a, axis=1, keepdims=True)
print(f'sum_student(axis=1):\n{sum_student}, shape: {sum_student.shape}')
```

### 1-5.Mean Subtraction

```python
import numpy as np

n_student, n_class = 3,4
m_score, M_score = 0, 100
scores = np.random.randint(low=m_score, high=M_score,
                           size = (n_student,n_class))

print(f'scores:\n{scores}\n')

mean_class = np.mean(scores, axis=0, keepdims=True)
print(f'mean_class(axis=0):\n{mean_class}, shape: {mean_class.shape}')

mean_student = np.mean(scores, axis=1, keepdims=True)
print(f'mean_student(axis=1):\n{mean_student}, shape: {mean_student.shape}\n')

print(f'Mean Subtraction by classes:\n{scores - mean_class}')
print(f'Mean Subtraction by students:\n{scores - mean_student}')
```

## 2.텐서 ndarray에서의 axis, keepdims

### 2-1.axis, keepdims

```python
import numpy as np

a = np.arange(24).reshape((2,3,-1))
print(f'ndarray:\n{a}, shape: {a.shape}\n')

sum = a.sum(axis=0)
print(f'sum(axis=0):\n{sum}, shape: {sum.shape}')
sum = a.sum(axis=0, keepdims=True)
print(f'sum(axis=0, keepdims=True):\n{sum}, shape: {sum.shape}\n')

sum = a.sum(axis=1)
print(f'sum(axis=1):\n{sum}, shape: {sum.shape}')
sum = a.sum(axis=1, keepdims=True)
print(f'sum(axis=1, keepdims=True):\n{sum}, shape: {sum.shape}\n')

sum = a.sum(axis=2)
print(f'sum(axis=2):\n{sum}, shape: {sum.shape}')
sum = a.sum(axis=2, keepdims=True)
print(f'sum(axis=2, keepdims=True):\n{sum}, shape: {sum.shape}\n')
```

### 2-2.Example

```python
import numpy as np

n_test_time, n_student, n_class = 4, 3, 4
m_score, M_score = 0, 100
scores = np.random.randint(low=m_score, high=M_score,
                           size = (n_test_time, n_student, n_class))

print(f'scores:\n{scores}, shape: {scores.shape}\n')

test_mean = np.mean(scores, axis=0)
print(f'test_mean:\n{test_mean}, shape: {test_mean.shape}\n')

class_mean = np.mean(scores, axis=1, keepdims=True)
print(f'class_mean:\n{class_mean}, shape: {class_mean.shape}\n')

student_mean = np.mean(scores, axis=2, keepdims=True)
print(f'student_mean:\n{student_mean}, shape: {student_mean.shape}')
```

```python
import numpy as np

C,H,W = 3,100,200

# (C,H,W) case
image = np.random.randint(0, 256, size =(C,H,W))
print(f'Shape of original image: {image.shape}')

gray_image = np.mean(image, axis=0)
print(f'Shape of gray-scaled image: {gray_image.shape}\n')

# (H,W,C) case
image = np.random.randint(0, 256, size =(H,W,C))
print(f'Shape of original image: {image.shape}')

gray_image = np.mean(image, axis=-1, keepdims=True) # axis=-1: 마지막 차원
print(f'Shape of gray-scaled image: {gray_image.shape}')
```
