## 1.상수(Constants)

```python
import numpy as np

PI = np.pi # 원주율
E = np.e   # 자연상수

print(f'pi: {PI}')
print(f'natural constant: {E}')
```

## 2.Trigonometric Functions(삼각함수)

### 2-1.deg2rad and rad2deg
- np.rad2deg(x)
- np.deg2rad(x)

```python
import numpy as np

degree = np.array([30, 45, 60, 90, 180, 360])
rad = np.deg2rad(degree)
degree = np.rad2deg(rad)

print(f'radian: {rad.round(2)}')
print(f'degree: {degree}')
```

### 2-2. Trigonometric Functions
- np.sin(x)
- np.cos(x)
- np.tan(x)

```python
import numpy as np

x = np.deg2rad(np.linspace(0, 360, 11))
sin, cos, tan = np.sin(x), np.cos(x), np.tan(x)

print(f'np.sin(x):\n{sin.round(2)}')
print(f'np.cos(x):\n{cos.round(2)}')
print(f'np.tan(x):\n{tan.round(2)}')
print(f'np.sin(x)/np.cos(x):\n{(sin/cos).round(2)}')
```

### 2-3.Graph

```python
import numpy as np
import matplotlib.pyplot as plt

PI = np.pi
x = np.linspace(0, 4*PI, 100)

sin, cos, tan = np.sin(x), np.cos(x), np.tan(x)

plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, sin, label=r'$y = sin(x)$')
ax.plot(x, cos, label=r'$y = cos(x)$')
ax.plot(x, tan, label=r'$y = tan(x)$')

xticks = np.arange(0, 4*PI+0.1, 0.5*PI)
xticklabels = [str(xtick) + r'$\frac{\pi}{2}$' for xtick in range(9)]
ax.set_xticks(xticks)
ax.set_xticklabels(xticklabels)
ax.tick_params(axis ='x',labelsize=20)
ax.set_ylim([-2, 2])
ax.legend()
```

## 3.Hyperbolic Functions(쌍곡선 함수)

### 3-1.Exponential Functions
- np.exp(x)

```python
import numpy as np

E = np.e
x = np.arange(1,7)

print(f'E**x:\n{(E**x).round(2)}')
print(f'np.exp(x):\n{np.exp(x).round(2)}')
```

```python
# sigmoid Function
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5,5, 100)
sigmoid = 1/(1+np.exp(-x))

plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x, sigmoid)
ax.tick_params(labelsize=20)
```

### 3-2.Hyperbolic Functions
- np.sinh(x)
- np.cosh(x)
- np.tanh(x)

```python
import numpy as np

x = np.linspace(0, 1, 5)
sinh, cosh, tanh = np.sinh(x), np.cosh(x), np.tanh(x)

print(f'np.sinh(x):\n{sinh.round(2)}')
print(f'np.cosh(x):\n{cosh.round(2)}')
print(f'np.tanh(x):\n{tanh.round(2)}')
print(f'np.sinh(x)/np.cosh(x):\n{(sinh/cosh).round(2)}')
```

```python
import numpy as np

x = np.linspace(0, 1, 5)

sinh = np.sinh(x)
sinh_exp = (np.exp(x) - np.exp(-x))/2
print(f'sinh: {sinh.round(2)}')
print(f'sinh_exp: {sinh_exp.round(2)}\n')

cosh = np.cosh(x)
cosh_exp = (np.exp(x) + np.exp(-x))/2
print(f'cosh: {cosh.round(2)}')
print(f'cosh_exp: {cosh_exp.round(2)}\n')

tanh = np.tanh(x)
tanh_exp = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
print(f'tanh: {tanh.round(2)}')
print(f'tanh_exp: {tanh_exp.round(2)}')
```

### 3-3.Activation Functions


```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-3, 3, 100)

sigmoid = 1/(1 + np.exp(-x))
tanh = (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
relu = np.where(x<0, 0, x)

plt.style.use('seaborn')
fig, ax = plt.subplots(figsize=(10,5))
ax.plot(x,sigmoid, label=r'$y = \sigma(x)$')
ax.plot(x,tanh, label=r'$y = tanh(x)$')
ax.plot(x,relu, label=r'$y = Relu(x)$')

ax.tick_params(labelsize=20)
ax.legend(fontsize=20)
```

## 4. Quadratic, Irrational, Rational Functions

### 4-1.Quadratic Funtions(2차 함수)
- np.square(x)

```python
import numpy as np
a = np.random.randint(0,10, (10, ))
print(f'a:\n{a}\n')

square1 = a*a
square2 = a**2
square3 = np.square(a)

print(f'a*a:\n{square1}')
print(f'a**2:\n{square2}')
print(f'np.sqaure(a):\n{square3}')
```

### 4-2.Irrational Functions(무리 함수)
- np.sqrt(x)
- np.cbrt(x)

```python
import numpy as np

a = np.random.randint(0, 10, (4, ))
print(f'a:\n{a}\n')

sqrt1 = a**(1/2)
sqrt2 = np.sqrt(a)
print(f'a**(1/2):\n{sqrt1.round(2)}')
print(f'np.sqrt(a):\n{sqrt2.round(2)}\n')

cbrt1 = a**(1/3)
cbrt2 = np.cbrt(a)
print(f'a**(1/3):\n{cbrt1.round(2)}')
print(f'np.cbrt(a):\n{cbrt2.round(2)}\n')
```

### 4-3.Rational Functions(분수 함수)
- np.reciprocal(x)

```python
import numpy as np

a = np.random.uniform(0, 10, (4,))
print(f'a:\n{a.round(2)}\n')

recip1 = 1/a
recip2 = a**(-1)
recip3 = np.reciprocal(a)

print(f'1/a:\n{recip1.round(2)}')
print(f'a**(-1):\n{recip2.round(2)}')
print(f'np.reciprocal(a):\n{recip3.round(2)}')
```

## 5.Power Functions(멱함수)
- np.power(x1, x2)

```python
import numpy as np

a = np.random.uniform(0, 5, (4, ))
print(f'a:\n{a.round(2)}\n')

s1 = a**2
s2 = np.square(a)
s3 = np.power(a, 2)
print(f'a**2:\n{s1.round(2)}')
print(f'np.square(a):\n{s2.round(2)}')
print(f'np.power(a, 2):\n{s3.round(2)}\n')

re1 = a**(-1)
re2 = np.reciprocal(a)
re3 = np.power(a, -1)
print(f'a**(-1):\n{re1.round(2)}')
print(f'np.reciporcal(a):\n{re2.round(2)}')
print(f'np.power(a, -1):\n{re3.round(2)}\n')

sqrt1 = a**(1/2)
sqrt2 = np.sqrt(a)
sqrt3 = np.power(a, 1/2)
print(f'a**(1/2):\n{sqrt1.round(2)}')
print(f'np.sqrt(a):\n{sqrt2.round(2)}')
print(f'np.power(a, 1/2):\n{sqrt3.round(2)}\n')

cbrt1 = a**(1/3)
cbrt2 = np.cbrt(a)
cbrt3 = np.power(a, 1/3)
print(f'a**(1/3):\n{cbrt1.round(2)}')
print(f'np.cbrt(a):\n{cbrt2.round(2)}')
print(f'np.power(a, 1/3):\n{cbrt3.round(2)}\n')

exp1 = np.e**a
exp2 = np.exp(a)
exp3 = np.power(np.e, a)
print(f'np.e**a:\n{exp1.round(2)}')
print(f'np.exp(a):\n{exp2.round(2)}')
print(f'np.power(np.e, a):\n{exp3.round(2)}\n')

b = np.random.uniform(0, 5, (4, ))
print(f'b:\n{b.round(2)}')
power1 = a**b
power2 = np.power(a, b)
print(f'a**b:\n{power1.round(2)}')
print(f'np.power(a, b):\n{power2.round(2)}')
```

## 6.Log Functions(로그 함수)

### 6-2.역함수 관계


```python
import numpy as np

a = np.random.uniform(1,5,(4, ))
print(f'a:\n{a.round(2)}\n')

log = np.log(a)
exp = np.exp(log)
print(f'log:\n{log.round(2)}')
print(f'exp:\n{exp.round(2)}')
```

### 6-3.Properties of Log

```python
# log(ab) = log(a) + log(b)
import numpy as np

a = np.random.uniform(1, 5, (4, ))
b = np.random.uniform(1, 5, (4, ))

print(f'np.log(a*b):\n{np.log(a*b).round(2)}')
print(f'np.log(a) + np.log(b):\n{(np.log(a) + np.log(b)).round(2)}\n')

# loga(b) = logc(b)/logc(a)

a = np.random.uniform(1, 5, (4, ))

log2 = np.log(a)/np.log(2)
print(f'log2(a):\n{log2}')
log3 = np.log(a)/np.log(3)
print(f'log3(a):\n{log3}')
```

### 6-4.Binary Entropy

```python
import numpy as np

p = np.random.uniform(0.001, 1, (4, ))
print(f'p:\n{p.round(2)}\n')

# 밑이 e
be_e = -(p*np.log(p) + (1-p)*np.log(1-p))
print(f'binary entropy(base e):\n{be_e.round(2)}')

# 밑이 2
be_2 = -(p*(np.log(p)/np.log(2)) + (1-p)*(np.log(1-p)/np.log(2)))
print(f'binary entropy(base 2):\n{be_2.round(2)}')
```
