## 1.from Python Lists

### 1-1.Making Vector ndarrays

```python
import numpy as np

vec_py = [1, 2, 3]
vec_np = np.array(vec_py)

print(f'vec_py = {vec_py}, type = {type(vec_py)}')
print(f'vec_np = {vec_np}, type = {type(vec_np)}')
```

### 1-2.Making Matrix ndarrays

```python
import numpy as np

mat_py = [[1,2,3],
          [4,5,6]]
mat_np = np.array(mat_py)

print(f'vec_py =\n{mat_py}, type = {type(mat_py)}')
print(f'vec_np =\n{mat_np}, type = {type(mat_np)}')
```

### 1-3.Making 3rd Order Tensor ndarrays

```python
import numpy as np

tensor_py = [[[1,2,3],
              [4,5,6]],
             
             [[11,12,13],
              [14,15,16]]]
tensor_np = np.array(tensor_py)

print(f'vec_py =\n{tensor_py}, type = {type(tensor_py)}')
print(f'vec_np =\n{tensor_np}, type = {type(tensor_np)}')
```

### 1-4. Shapes of ndarrays

```python
import numpy as np

scalar_np = np.array(3.14)
vec_np = np.array([1,2,3])
mat_np = np.array([[1,2,3],
                   [4,5,6]])
tensor_np = np.array([[[1,2,3],
                       [4,5,6]],
                      
                      [[11,12,13],
                       [14,15,16]]])

print(f'shape of scalar_np: {scalar_np.shape}')
print(f'shape of vec_np: {vec_np.shape}')
print(f'shape of mat_np: {mat_np.shape}')
print(f'shape of tensor_np: {tensor_np.shape}\n')

print(f'dimension of scalar_np: {len(scalar_np.shape)}')
print(f'dimension of vec_np: {len(vec_np.shape)}')
print(f'dimension of mat_np: {len(mat_np.shape)}')
print(f'dimension of tensor_np: {len(tensor_np.shape)}')
```

## 2.with Shapes

### 2-1.ndarrays with Specific Values

```python
import numpy as np

# --- np.zeros(shape, dtype=float, order='C', *, like=None)
M = np.zeros(shape=(2,3))

print(f'shape of M: {M.shape}')
print(f'M =\n{M}\n')
# ---

# --- np.ones(shape, dtype=float, order='C', *, like=None)
M = np.ones(shape=(2,3))

print(f'shape of M: {M.shape}')
print(f'M =\n{M}\n')
# ---

# --- np.full(shape, fill_value, dtype=None, order='C', *, like=None)
M = np.full(shape=(2,3), fill_value=3.14)

print(f'shape of M: {M.shape}')
print(f'M =\n{M}\n')
# ---

# --- np.empty(shape, dtype=float, order='C', *, like=None) : 메모리에 빈공간을 잡아두는 메서드
M = np.empty(shape=(3,3))

print(f'shape of M: {M.shape}')
print(f'M =\n{M}\n')
# ---
```

### 2-2.from Existing ndarrays

```python
import numpy as np

M = np.full(shape=(2,3), fill_value=3.14)
print(f'shape of M: {M.shape}')
print(f'M =\n{M}\n')

# --- np.zeros_like(a, dtype=None, order='K', subok=True, shape=None)
zeros_like = np.zeros_like(M)

print(f'shape of zeoros_like(M): {zeros_like.shape}')
print(f'zeros_like(M) =\n{zeros_like}\n')
# ---

# --- np.ones_like(a, dtype=None, order='K', subok=True, shape=None)
ones_like = np.ones_like(M)

print(f'shape of ones_like(M): {ones_like.shape}')
print(f'ones_like(M) =\n{ones_like}\n')
# ---

# --- np.full_like(a, fill_value, dtype=None, order='K', subok=True, shape=None)
full_like = np.full_like(M, fill_value = 100)

print(f'shape of full_like(M): {full_like.shape}')
print(f'full_like(M) =\n{full_like}\n')
# ---

# --- np.empty_like(prototype, dtype=None, order='K', subok=True, shape=None)
empty_like = np.empty_like(M)

print(f'shape of empty_like(M): {empty_like.shape}')
print(f'empty_like(M) =\n{empty_like}\n')
# ---
```

## 3.with the Fixed Intervals/Points

### 3-1.with Fixed Intervals
np.arange([start], stop, [step], dtype=None, *, like=None)

```python
import numpy as np

print(f'np.arange(10) = {np.arange(10)}, type: {type(np.arange(10))}')
print(f'np.arange(2,5) = {np.arange(2,5)}, type: {type(np.arange(2,5))}')
print(f'np.arange(2,10,2) = {np.arange(2,10,2)}, type: {type(np.arange(2,10,2))}')

print(f'\nnp.arange(10.5) = {np.arange(10.5)}, type: {type(np.arange(10.5))}')
print(f'np.arange(1.5,10.5) = {np.arange(1.5,10.5)}, type: {type(np.arange(1.5,10.5))}')
print(f'np.arange(1.5,10.5,2.5) = {np.arange(1.5,10.5,2.5)}, type: {type(np.arange(1.5,10.5,2.5))}')
```

### 3-2.with Fixed Points
np.linspace(start, stop, num=50, endpoint=True, retstep=False, dtype=None, axis=0)

```python
import numpy as np

print(f'np.linspace(0,1,5) = {np.linspace(0,1,5)}, type: {type(np.linspace(0,1,5))}\n')

a = np.linspace([1,10,100], [2,20,200], 5)
print(f'np.linspace([1,10,100], [2,20,200], 5) =\n{a}, type: {type(a)}\n')
```

## 4.from Random Distributions

### 4-1.from Normal Distributions

```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn')
fig, axes = plt.subplots(2,1,figsize=(10,10))

# --- np.random.randn(d0, d1, ..., dn) -> 평균=0, 표준편차=1(Standard Normal Distribution)
random_values = np.random.randn(3000)
print(f'shape of random_values: {random_values.shape}')
axes[0].hist(random_values,bins=30)
axes[0].set_title('mean=0, std = 1',fontsize=20)
print('------------------------------------\n')
# ---

# --- np.random.normal(loc=0.0, scale=1.0, size=None) -> 평균(loc), 표준편차(scale) 지정 가능
normal_values = np.random.normal(loc=-2, scale=1, size=(3000))
print(f'shape of normal_values: {normal_values.shape}\n')
axes[1].hist(normal_values, bins=30)
axes[1].set_title('mean=-2, std=1',fontsize=20)

normal = np.random.normal(loc = [-2,0,3],
                          scale = [1,2,5],
                          size=(200,3))
print(f'shape of normal: {normal.shape}\n')

print(f'np.random.normal(loc=-2, scale=1, size=(3,3)) =\n{np.random.normal(loc=-2, scale=1, size=(3,3))}')
# ---
plt.show()
```

### 4-2.from Uniform Distributions

```python
import numpy as np

# --- random.rand(d0, d1, d2, ..., dn) -> [0,1)
rand = np.random.rand(4,4)
print(f'np.random.rand(4,4) =\n{rand}, shape: {rand.shape}')
print('---------------------------------------------------\n')
# ---

# --- random.uniform(low=0.0, high=1.0, size=None) -> [low, high)
uniform = np.random.uniform(low=-10, high=10, size=(4,4))
print(f'np.random.uniform(low=-10, high=10, size=(4,4)) =\n{uniform}, shape: {uniform.shape}')
print('---------------------------------------------------\n')
# ---

# --- random.randint(low =None, high=None, size=None, dtype=int) -> [low, high)
randint = np.random.randint(low=0, high=7, size=(4,4))
print(f'np.random.randint(low=0, high=7, size=(4,4)) =\n{randint}, shape: {randint.shape}')
# ---
```
