## 1.원소별로 ndarray 반복

### 1-1.np.repeat(a, repeats, axis=None)

```python
# Scalar
import numpy as np

x = 3
print(f'x: {x}\n')

rep = np.repeat(x, 5)
print(f'np.repeat(x, 5): {rep}')
```

```python
# Vector
import numpy as np

x = np.array([1, 2, 3])
print(f'x: {x}\n')

rep0 = np.repeat(x, 3)
print(f'np.repeat(x, 3): {rep0}')
```

### 1-2.np.repeat with axis

```python
import numpy as np

x = np.arange(4).reshape((2,2))
print(f'x:\n{x}\n')

rep = np.repeat(x, 3)
print(f'np.repeat(x, 3):\n{rep}\n')

rep0 = np.repeat(x, 3, axis=0)
print(f'np.repeat(x, 3, axis=0):\n{rep0}\n')

rep1 = np.repeat(x, 3, axis=1)
print(f'np.repeat(x, 3, axis=1):\n{rep1}')
```

```python
import numpy as np

x = np.arange(4).reshape((2,2))
print(f'x:\n{x}\n')

rep = np.repeat(x, repeats=[2, 1], axis=0) # 첫번째 원소 두번 반복, 두번째 원소 한번 반복
print(f'np.repeat(x, repeats=[2, 1], axis=0):\n{rep}\n')

rep = np.repeat(x, repeats=[1, 2], axis=0) # 첫번째 원소 한번 반복, 두번째 원소 두번 반복
print(f'np.repeat(x, repeats=[1, 2], axis=0):\n{rep}')

print('------------------------------------------------')

x = np.arange(6).reshape((2,3))
print(f'x:\n{x}\n')

rep = np.repeat(x, repeats=[2, 1,3], axis=1) 
print(f'np.repeat(x, repeats=[2, 1, 3], axis=1):\n{rep}\n')

rep = np.repeat(x, repeats=[1, 2,1], axis=1) 
print(f'np.repeat(x, repeats=[1, 2, 1], axis=1):\n{rep}')
```

### 1-3.np.repeat with Vectors

```python
# Row vector
import numpy as np

row_vec = np.arange(4).reshape((1, -1))
print(f'ndarray:\n{row_vec}, shape: {row_vec.shape}\n')

rep = np.repeat(row_vec, 3, axis=0)
print(f'np.repeat(row_vec, 3, axis=0):\n{rep}, shape: {rep.shape}')

print('------------------------------------------------')

# Column vector
col_vec = np.arange(4).reshape((-1, 1))
print(f'ndarray:\n{col_vec}, shape: {col_vec.shape}\n')

rep = np.repeat(col_vec, 3, axis=1)
print(f'np.repeat(col_vec, 3, axis=1):\n{rep}, shape: {rep.shape}')
```

## 2.전체적으로 ndarray 반복시키기

### 2-1.np.tile(A, reps)

```python
# Vector
import numpy as np

a = np.arange(4)
print(f'ndarray:\n{a}, shape: {a.shape}\n')

tile = np.tile(a, 2)
print(f'np.tile(a, 2):\n{tile}, shape: {tile.shape}\n')

tile = np.tile(a, reps=[1, 2]) # 0차원 한번 반복, 1차원 두번 반복
print(f'np.tile(a, reps=[1, 2]):\n{tile}, shape: {tile.shape}\n')

tile = np.tile(a, reps=[2, 1]) # 0차원 두번 반복, 1차원 한번 반복
print(f'np.tile(a, reps=[2, 1]):\n{tile}, shape: {tile.shape}\n')

tile = np.tile(a, reps=[2, 2]) 
print(f'np.tile(a, reps=[2, 2]):\n{tile}, shape: {tile.shape}')
```

```python
# Matrix
import numpy as np

a = np.arange(6).reshape((2, 3))
print(f'ndarray:\n{a}, shape: {a.shape}\n')

tile = np.tile(a, reps=[1, 2])
print(f'np.tile(a, reps=[1, 2]):\n{tile}, shape: {tile.shape}\n')

tile = np.tile(a, reps=[2, 1])
print(f'np.tile(a, reps=[2, 1]):\n{tile}, shape: {tile.shape}\n')

tile = np.tile(a, reps=[2, 2])
print(f'np.tile(a, reps=[2, 2]):\n{tile}, shape: {tile.shape}')
```

### 2-2.np.tile with Vectors

```python
# Row vector
import numpy as np

a = np.arange(4).reshape((1, -1))
print(f'ndarray:\n{a}, shape: {a.shape}\n')

tile = np.tile(a, reps=[5, 1])
print(f'np.tile(a, reps=[5, 1]):\n{tile}, shape: {tile.shape}')

print('------------------------------------------------')

a = np.arange(4).reshape((-1, 1))
print(f'ndarray:\n{a}, shape: {a.shape}\n')

tile = np.tile(a, reps=[1, 5])
print(f'np.tile(a, reps=[1, 5]):\n{tile}, shape: {tile.shape}')
```

## 3.ndarray 반복의 응용

### 3-1.Row-wise Case

```python
import numpy as np

x = np.arange(12).reshape((4, 3))
print(f'x:\n{x}, shape: {x.shape}\n')

x = x.sum(axis=0, keepdims=True)
print(f'x.sum:\n{x}, shape: {x.shape}\n')

x = x.repeat(3, axis=0)
print(f'x.repeat:\n{x}, shape: {x.shape}\n')

x = x.T
print(f'x.T:\n{x}, shape: {x.shape}')

print('------------------------------------------------')

# 위의 것을 한줄로 표현
x = np.arange(12).reshape((4, 3))
print(f'x:\n{x}, shape: {x.shape}\n')

y = x.sum(axis=0, keepdims=True).repeat(3, axis=0).T
print(f'y:\n{y}, shape: {y.shape}')
```

### 3-2.Column-wise Case

```python
import numpy as np

x = np.arange(12).reshape((4, 3))
print(f'x:\n{x}, shape: {x.shape}\n')

y = x.sum(axis=1, keepdims=True).repeat(2, axis=1).T
print(f'y:\n{y}, shape: {y.shape}')
```

## 4.Making coordinates(좌표축 만들기)

### 4-1.Coordinates from 1D Vectors

```python
import numpy as np

x = np.arange(-2, 3)
y = np.arange(-2, 3)
print(f'x: {x}')
print(f'y: {y}\n')

X = x.reshape((1, -1)).repeat(y.shape[0], axis=0)
Y = y.reshape((-1, 1)).repeat(x.shape[0], axis=1)
print(f'X:\n{X}')
print(f'Y:\n{Y}\n')

Z = np.square(X) + np.square(Y)
print(f'Z:\n{Z}')
```

### 4-2.np.meshgrid

```python
import numpy as np

x = np.arange(-2, 3)
y = np.arange(-2, 3)
print(f'x: {x}')
print(f'y: {y}\n')

X,Y = np.meshgrid(x, y)
print(f'X:\n{X}')
print(f'Y:\n{Y}\n')

Z = np.square(X) + np.square(Y)
print(f'Z:\n{Z}')
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
print(f'x.shape: {x.shape}')
print(f'y.shape: {y.shape}\n')

X, Y = np.meshgrid(x, y)
print(f'X.shape: {X.shape}')
print(f'Y.shape: {Y.shape}')

Z = np.square(X) + np.square(Y)

fig = plt.figure(figsize=(7,7))
ax = fig.add_subplot(projection = '3d')

ax.plot_wireframe(X, Y, Z)
ax.tick_params(labelsize=20)
```
