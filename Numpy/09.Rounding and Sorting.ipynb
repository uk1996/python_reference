## 1.Rounding Functions


### 1-1.반올림 API
- np.around(a, decimals=0, out=None)
- np.round_(a, decimals=0, out=None)
- ndarray.round(decimals=0, out=None)

```python
import numpy as np

x = np.random.uniform(-5,5, (5,))
print(f'x:\n{x}\n')

np_around = np.around(x, decimals=2)
np_round_ = np.round_(x, decimals=2)
x_round = x.round(decimals=2)

print(f'np_around:\n{np_around}')
print(f'np_round_:\n{np_round_}')
print(f'x_round:\n{x_round}')
```

```python
import numpy as np

scores = np.random.uniform(0,100, (100,5))

means = np.mean(scores, axis=0)
stds = np.std(scores, axis=0)

print(f'calss means:\n{means}')
print(f'class stds:\n{stds}\n')

print(f'calss means:\n{means.round(2)}')
print(f'class stds:\n{stds.round(2)}\n')
```

### 1-2.올림,버림 API
- np.ceil(x, /, out=None, *, where=True, casting='same_kind', order='K', dtype=None, subok=True[,signature,extobj])
- np.floor(x, /, out=None, *, where=True, casting='same_kind', order='K', dtype=None, subok=True[,signature,extobj])

```python
import numpy as np

x = np.random.uniform(-5,5, (5,))
print(f'x:\n{x}\n')

ceil = np.ceil(x)
floor = np.floor(x)
print(f'ceil:\n{ceil}')   # 올림
print(f'floor:\n{floor}') # 버림
```

### 1-3.Truncation Functions
np.trunc(x, /, out=None, *, where=True, casting='same_kind', order='K', dtype=None, subok=True[,signature,extobj])

```python
import numpy as np

x = np.random.uniform(-5,5, (5,))
print(f'x:\n{x}\n')

trunc = np.trunc(x)
print(f'trunc:\n{trunc}')
```

```python
import numpy as np

x = np.random.uniform(-5,5, (5,))
print(f'x:\n{x}\n')

trunc = np.trunc(x)
trunc_where = np.where(x >= 0, np.floor(x), np.ceil(x))
print(f'trunc:\n{trunc}')
print(f'trunc_where:\n{trunc_where}')
```

```python
import numpy as np

x = np.random.uniform(-5,5, (5,))
print(f'x:\n{x}\n')

# 소수점 둘째자리까지 나타내기
trunc = 0.01*np.trunc(100*x)
print(f'trunc:\n{trunc}')
```

```python
import numpy as np

x = np.random.uniform(-5,5, (5,))
print(f'x:\n{x}\n')

# 정수 부분, 소수 부분
int_part = np.trunc(x)
frac_part = x - int_part
print(f'int_part:\n{int_part}')
print(f'frac_part:\n{frac_part}')
```

## 2.Sorting Functions
- np.sort(a, axis=-1, kind=None, order=None)
- np.argsort(a, axis=-1, kind=None, order=None)

### 2-1.Sorting Vectors

```python
import numpy as np

x = np.random.randint(0,100, (10,))
print(f'x:\n{x}\n')

# 오름차순(ascending)
sort = np.sort(x)
argsort = np.argsort(x)
print(f'sort:\n{sort}')
print(f'argsort:\n{argsort}\n')

# 내림차순(descending)
sort = np.sort(x)[::-1]
argsort = np.argsort(x)[::-1]
print(f'sort:\n{sort}')
print(f'argsort:\n{argsort}')
```

```python
# top3-Accuracy
import numpy as np

pred = np.random.uniform(0,100, (5,))
pred /= pred.sum()
print(f'pred:\n{pred.round(3)}\n')

top3_pred = np.sort(pred)[::-1][:3]
top3_indices = np.argsort(pred)[::-1][:3]
print(f'top3_pred:\n{top3_pred.round(3)}')
print(f'top3_indices:\n{top3_indices}')
```

### 2-2.Sorting Matrics

```python
# 오름차순(ascending)
import numpy as np

x = np.random.randint(0,100, (4,5))
print(f'x:\n{x}\n')

sort = np.sort(x, axis=0)
argsort = np.argsort(x, axis=0)
print(f'sort(axis=0):\n{sort}')
print(f'argsort(axis=0):\n{argsort}\n')

sort = np.sort(x, axis=1)
argsort = np.argsort(x, axis=1)
print(f'sort(axis=1):\n{sort}')
print(f'argsort(axis=1):\n{argsort}')
```

```python
# 내림차순(descending)
import numpy as np

x = np.random.randint(0,100, (4,5))
print(f'x:\n{x}\n')

sort = np.sort(x, axis=0)[::-1, :]
argsort = np.argsort(x, axis=0)[::-1, :]
print(f'sort(axis=0):\n{sort}')
print(f'argsort(axis=0):\n{argsort}\n')

sort = np.sort(x, axis=1)[:, ::-1]
argsort = np.argsort(x, axis=1)[:, ::-1]
print(f'sort(axis=1):\n{sort}')
print(f'argsort(axis=1):\n{argsort}')
```

```python
# top2-Students
import numpy as np

scores = np.random.randint(0,100, (5,3))
print(f'scores:\n{scores}\n')

sort = np.sort(scores, axis=0)[::-1, :]
argsort = np.argsort(scores, axis=0)[::-1, :]
print(f'sort:\n{sort}')
print(f'argsort:\n{argsort}\n')

print(f'top2-scores:\n{sort[:2, :]}')
print(f'top2-students:\n{argsort[:2, :]}')
```

```python
# bottom2-Classes
import numpy as np

scores = np.random.randint(0,100, (5,3))
print(f'scores:\n{scores}\n')

sort = np.sort(scores, axis=1)
argsort = np.argsort(scores, axis=1)
print(f'sort:\n{sort}')
print(f'argsort:\n{argsort}\n')

print(f'bottom2-scores:\n{sort[:, :2]}')
print(f'bottom2-classes:\n{argsort[:, :2]}')
```
