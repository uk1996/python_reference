## 1.Fully-connected Operation of Scalar

### 1-1.with For loop

```python
#method 1
import numpy as np

x = np.arange(5)
y = np.arange(2, 6)
print(f'x: {x}')
print(f'y : {y}\n')


for x_val in x:
    for y_val in y:
       print(f'{x_val} + {y_val} = {x_val + y_val},', end = ' ')
    print()

print('---------------------------------------------')

# method 2
x = np.arange(5)
y = np.arange(2, 6)
print(f'x: {x}')
print(f'y : {y}\n')

for x_val in x:
    print(f'{x_val} + {y} = {x_val + y}')
```

### 1-2.with np.meshgrid

```python
import numpy as np

x = np.arange(5)
y = np.arange(2, 6)
print(f'x: {x}')
print(f'y : {y}\n')

X, Y = np.meshgrid(x, y)
X, Y = X.T, Y.T
print(f'X:\n{X}')
print(f'Y:\n{Y}\n')

Z = X + Y
print(f'Z:\n{Z}')
```

### 1-3.with Broadcasting

```python
# 두개 더할때
import numpy as np

x = np.arange(5)
y = np.arange(2, 6)
print(f'x: {x}')
print(f'y : {y}\n')

X = x.reshape((-1,1))
Y = y.reshape((1, -1))
print(f'X:\n{X}, shape: {X.shape}')
print(f'Y:\n{Y}, shape: {Y.shape}\n')

Z = X + Y
print(f'Z:\n{Z}, shape: {Z.shape}\n')

print(f'x[0] +y: {Z[0, :]}')
print(f'x[4] +y: {Z[4, :]}\n')

print(f'x +y[0]: {Z[:, 0]}')
print(f'x +y[2]: {Z[:, 2]}')
```

```python
# 세개 더할때
import numpy as np

x = np.arange(5)
y = np.arange(2, 6)
z = np.arange(4,7)
print(f'x: {x}, shape: {x.shape}')
print(f'y : {y}, shape: {y.shape}')
print(f'z : {z}, shape: {z.shape}\n')

x = np.expand_dims(x, axis=(1,2))
y = np.expand_dims(y, axis=(0,2))
z = np.expand_dims(z, axis=(0,1))
print(f'x:\n{x}, shape: {x.shape}')
print(f'y:\n{y}, shape: {y.shape}')
print(f'z:\n{z}, shape: {z.shape}\n')

print(f'x+y+z:\n{x+y+z}, shape: {(x+y+z).shape}')
```

## 2.Fully-connected Operation of Vector

### 2-1.with For loop

```python
import numpy as np

X = np.random.uniform(-5, 5, (4, 2))
Y = np.random.uniform(-5, 5, (3, 2))
print(f'X:\n{X.round(2)}, shape: {X.shape}')
print(f'Y:\n{Y.round(2)}, shape: {Y.shape}\n')

for x in X:
    for y in Y:
        print((x+y).round(2), end=' ')
    print()
```

### 2-2.with Broadcasting

```python
# 두개 더할때
import numpy as np

X = np.random.uniform(-5, 5, (4, 2))
Y = np.random.uniform(-5, 5, (3, 2))
print(f'X:\n{X.round(2)}, shape: {X.shape}')
print(f'Y:\n{Y.round(2)}, shape: {Y.shape}\n')

X = np.expand_dims(X ,axis=1)
Y = np.expand_dims(Y, axis=0)
print(f'X:\n{X.round(2)}, shape: {X.shape}')
print(f'Y:\n{Y.round(2)}, shape: {Y.shape}\n')

Z = X + Y
print(f'Z:\n{Z.round(2)}, shape: {Z.shape}\n')

print(f'X[0] + Y:\n{Z[0, :, :].round(2)}')
print(f'X[2] + Y:\n{Z[2, :, :].round(2)}\n')

print(f'X + Y[0]:\n{Z[:, 0, :].round(2)}')
print(f'X + Y[2]:\n{Z[:, 2, :].round(2)}\n')
```

```python
# 세개 더할때
import numpy as np

X = np.random.uniform(-5, 5, (4, 2))
Y = np.random.uniform(-5, 5, (3, 2))
Z = np.random.uniform(-5, 5, (2, 2))
print(f'X:\n{X.round(2)}, shape: {X.shape}')
print(f'Y:\n{Y.round(2)}, shape: {Y.shape}')
print(f'z:\n{Z.round(2)}, shape: {Z.shape}\n')

X = np.expand_dims(X ,axis=(1, 2))
Y = np.expand_dims(Y, axis=(0, 2))
Z = np.expand_dims(Z, axis=(0, 1))
print(f'X:\n{X.round(2)}, shape: {X.shape}')
print(f'Y:\n{Y.round(2)}, shape: {Y.shape}')
print(f'Z:\n{Z.round(2)}, shape: {Z.shape}\n')

F = X + Y + X
print(f'F:\n{F.round(2)}, shape: {F.shape}')
```

### 2-3.Euclidean Distances

```python
# x,y 좌표
import numpy as np

X = np.random.uniform(-5, 5, (4, 2))
Y = np.random.uniform(-5, 5, (3, 2))
print(f'X:\n{X.round(2)}, shape: {X.shape}')
print(f'Y:\n{Y.round(2)}, shape: {Y.shape}\n')

# with For loop
e_distance = []
for x in X:
    e_distance_vec = []
    for y in Y:
        e_distance_vec.append(np.sqrt(np.sum(np.square(x-y))))
    e_distance.append(e_distance_vec)
e_distance = np.vstack(e_distance)

print(f'Euclidean Distance:\n{e_distance.round(2)}, shape: {e_distance.shape}')

print('-----------------------------------------------------------')       

# with Broadcasting
X = np.expand_dims(X, axis=1)
Y = np.expand_dims(Y, axis=0)
print(f'X:\n{X.round(2)}, shape: {X.shape}')
print(f'Y:\n{Y.round(2)}, shape: {Y.shape}\n')

e_distance = np.sqrt(np.sum(np.square(X - Y), axis=-1))
print(f'Euclidean Distance:\n{e_distance.round(2)}, shape: {e_distance.shape}')
```
