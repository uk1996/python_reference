## 1.ndim, shape, size

### 1-1.ndarray.ndim

```python
import numpy as np

scalar_np = np.array(3.14)
vector_np = np.array([1,2,3])
matrix_np = np.array([[1,2], 
                      [3,4]])
tensor_np = np.array([[[1,2,3],
                       [4,5,6]],
                      
                      [[11,12,13],
                       [14,15,16]]])

print(f'{scalar_np}, ndim: {scalar_np.ndim}\n')
print(f'{vector_np}, ndim: {vector_np.ndim}\n')
print(f'{matrix_np}, ndim: {matrix_np.ndim}\n')
print(f'{tensor_np}, ndim: {tensor_np.ndim}')
```

### 1-2.ndarray.shape

```python
import numpy as np

scalar_np = np.array(3.14)
vector_np = np.array([1,2,3])
matrix_np = np.array([[1,2], 
                      [3,4]])
tensor_np = np.array([[[1,2,3],
                       [4,5,6]],
                      
                      [[11,12,13],
                       [14,15,16]]])

print(f'{scalar_np}, shape: {scalar_np.shape}, ndim={len(scalar_np.shape)}\n')
print(f'{vector_np}, shape: {vector_np.shape}, ndim={len(vector_np.shape)}\n')
print(f'{matrix_np}, shape: {matrix_np.shape}, ndim={len(matrix_np.shape)}\n')
print(f'{tensor_np}, shape: {tensor_np.shape}, ndim={len(tensor_np.shape)}')
```

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([[1,2,3]])
c = np.array([[1],[2],[3]])

print(f'{a}, shape: {a.shape}\n')
print(f'{b}, shape: {b.shape}\n')
print(f'{c}, shape: {c.shape}')
```

### 2-2.ndarray.dtype

```python
import numpy as np

M = np.arange(10)
N = np.full(shape=(2,3), fill_value=3.14)

print(f'{M},dtype: {M.dtype}')
print(f'{N},dtype: {N.dtype}\n')

int_np = np.array([1,2,3])
float_np = np.array([1., 2., 3.])

print(f'{int_np}, dtype: {int_np.dtype}')
print(f'{float_np}, dtype: {float_np.dtype}')
```

```python
import numpy as np

int8_np = np.array([1,2,3], dtype=np.int8)
int16_np = np.array([1,2,3], dtype=np.int16)
int32_np = np.array([1,2,3], dtype=np.int32)
int64_np = np.array([1,2,3], dtype=np.int64)

print(f'int8_np dtype: {int8_np.dtype}')
print(f'int16_np dtype: {int16_np.dtype}')
print(f'int32_np dtype: {int32_np.dtype}')
print(f'int64_np dtype: {int64_np.dtype}\n')

uint8_np = np.array([1,2,3], dtype=np.uint8)
uint16_np = np.array([1,2,3], dtype=np.uint16)
uint32_np = np.array([1,2,3], dtype=np.uint32)
uint64_np = np.array([1,2,3], dtype=np.uint64)

print(f'uint8_np dtype: {uint8_np.dtype}')
print(f'uint16_np dtype: {uint16_np.dtype}')
print(f'uint32_np dtype: {uint32_np.dtype}')
print(f'uint64_np dtype: {uint64_np.dtype}\n')

float32_np = np.array([1,2,3], dtype=np.float32)
float64_np = np.array([1,2,3], dtype=np.float64)

print(f'float32_np dtype: {float32_np.dtype}')
print(f'float64_np dtype: {float64_np.dtype}')
```

```python
import numpy as np

int8_np = np.array([1.5, 2.5, 3.5], dtype = np.int8)
uint8_np = np.array([1.5, 2.5, 3.5], dtype = np.uint8)

print(f'int8_np: {int8_np}')
print(f'uint8_np: {uint8_np}\n')

M = np.ones(shape=(2,3), dtype=np.float32)
N = np.zeros_like(M, dtype=np.float64)

print(f'{M}, dtype: {M.dtype}\n')
print(f'{N}, dtype: {N.dtype}')
```

### 2-3.ndarray.itemsize

```python
int8_np = np.array([1,2,3], dtype=np.int8)
int16_np = np.array([1,2,3], dtype=np.int16)
int32_np = np.array([1,2,3], dtype=np.int32)
int64_np = np.array([1,2,3], dtype=np.int64)

print(f'int8_np dtype: {int8_np.dtype}, itemsize={int8_np.itemsize}B')
print(f'int16_np dtype: {int16_np.dtype}, itemsize={int16_np.itemsize}B')
print(f'int32_np dtype: {int32_np.dtype}, itemsize={int32_np.itemsize}B')
print(f'int64_np dtype: {int64_np.dtype}, itemsize={int64_np.itemsize}B\n')

uint8_np = np.array([1,2,3], dtype=np.uint8)
uint16_np = np.array([1,2,3], dtype=np.uint16)
uint32_np = np.array([1,2,3], dtype=np.uint32)
uint64_np = np.array([1,2,3], dtype=np.uint64)

print(f'uint8_np dtype: {uint8_np.dtype}, itemsize={uint8_np.itemsize}B')
print(f'uint16_np dtype: {uint16_np.dtype}, itemsize={uint16_np.itemsize}B')
print(f'uint32_np dtype: {uint32_np.dtype}, itemsize={uint32_np.itemsize}B')
print(f'uint64_np dtype: {uint64_np.dtype}, itemsize={uint64_np.itemsize}B\n')


float32_np = np.array([1,2,3], dtype=np.float32)
float64_np = np.array([1,2,3], dtype=np.float64)

print(f'float32_np dtype: {float32_np.dtype}, itemsize={float32_np.itemsize}B')
print(f'float64_np dtype: {float64_np.dtype}, itemsize={float64_np.itemsize}B')
```

```python
import numpy as np

normal = np.random.normal(size=(50,50,32,5))
print(f'size: {normal.size}')
print(f'dtype/itemsize: {normal.dtype} / {normal.itemsize}B\n')

m_cap = normal.size * normal.itemsize
print(f'Memory capacity in B: {m_cap}B')
print(f'Memory capacity in KB: {m_cap/1024}KB')
print(f'Memory capacity in MB: {m_cap/1024**2}MB')
```

### 2-4.ndarray.nbytes

```python
import numpy as np

normal = np.random.normal(size=(50,50,32,5))

print(f'normal.size * normal.itemsize = {normal.size * normal.itemsize}B')
print(f'normal.nbytes: {normal.nbytes}B')
```
