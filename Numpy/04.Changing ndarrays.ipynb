## 1.np.reshape and ndarray.reshape


### 1-1.np.reshape(a, new_shape, order='C')

```python
import numpy as np

# --- np.reshape(a, newshape, order='C')
a = np.arange(6)
b = np.reshape(a, (2,3))
print(f'original ndarray:\n{a}')
print(f'reshaped ndarray:\n{b}\n')

a = np.arange(24)
b = np.reshape(a, (2,3,4))
print(f'original ndarray:\n{a}')
print(f'reshaped ndarray:\n{b}\n')
```

### 1-2.ndarray.reshape(shape, order='C')

```python
import numpy as np

a = np.arange(6)
b = a.reshape((2,3))
print(f'original ndarray:\n{a}')
print(f'reshaped ndarray:\n{b}\n')

a = np.arange(24)
b = a.reshape((2,3,4))
print(f'original ndarray:\n{a}')
print(f'reshaped ndarray:\n{b}\n')
```

### 1-3. the difference between np.reshape and ndarray.reshape

```python
a = np.random.randint(0,100,(100,))

# np.reshape
print(np.max(np.mean(np.reshape(a,(20,5)), axis=0)))

# ndarray.reshape
print(a.reshape((20,5)).mean(axis=0).max())
```

### 1-4.-1 in np.reshape
나머지 차원의 값이 정해지면 남은 한차원의 고정된값을 자동적으로 연산

```python
 a = np.arange(12)
 b = a.reshape((2,-1))
 c = a.reshape((3,-1))
 d = a.reshape((4,-1))
 e = a.reshape((6,-1))

print(f'original ndarray:\n{a}, shape = {a.shape}\n')
print(f'reshaped ndarray1:\n{b}, shape = {b.shape}\n')
print(f'reshaped ndarray2:\n{c}, shape = {c.shape}\n')
print(f'reshaped ndarray3:\n{d}, shape = {d.shape}\n')
print(f'reshaped ndarray4:\n{e}, shape = {e.shape}')
```

```python
 a = np.arange(12)
 b = a.reshape((-1,2))
 c = a.reshape((-1,3))
 d = a.reshape((-1,4))
 e = a.reshape((-1,6))

print(f'original ndarray:\n{a}, shape = {a.shape}\n')
print(f'reshaped ndarray1:\n{b}, shape = {b.shape}\n')
print(f'reshaped ndarray2:\n{c}, shape = {c.shape}\n')
print(f'reshaped ndarray3:\n{d}, shape = {d.shape}\n')
print(f'reshaped ndarray4:\n{e}, shape = {e.shape}')
```

```python
a = np.arange(24)
b = a.reshape((2,3,-1))
c = a.reshape((2,-1,4))
d = a.reshape((-1,3,4))

print(f'original ndarray:\n{a}, shape = {a.shape}\n')
print(f'reshaped ndarray1:\n{b}, shape = {b.shape}\n')
print(f'reshaped ndarray2:\n{c}, shape = {c.shape}\n')
print(f'reshaped ndarray3:\n{d}, shape = {d.shape}')
```

```python
a = np.random.randint(0,10,(2,2))
row_vector = a.reshape((1,-1))
col_vector = a.reshape((-1,1))

print(f'original ndarray:\n{a}, shape = {a.shape}\n')
print(f'reshaped ndarray1:\n{row_vector}, shape = {row_vector.shape}\n')
print(f'reshaped ndarray2:\n{col_vector}, shape = {col_vector.shape}')
```

## 2.np.resize and ndarray.resize

### 2-1.np.resize(a, new_shape)

```python
import numpy as np

a = np.arange(6)
b = np.resize(a, (2,3))
print(f'original ndarray:\n{a}')
print(f'resized ndarray:\n{b}\n')

a = np.arange(6)
b = np.resize(a, (9,))
print(f'original ndarray:\n{a}')
print(f'resized ndarray:\n{b}\n')

a = np.arange(6)
b = np.resize(a, (4))
print(f'original ndarray:\n{a}')
print(f'resized ndarray:\n{b}\n')

a = np.arange(6)
b = np.resize(a, (3,4))
print(f'original ndarray:\n{a}')
print(f'resized ndarray:\n{b}\n')

a = np.arange(9)
b = np.resize(a, (2,3,3))
print(f'original ndarray:\n{a}')
print(f'resized ndarray:\n{b}\n')

a = np.arange(9)
b = np.resize(a, (2,2))
print(f'original ndarray:\n{a}')
print(f'resized ndarray:\n{b}')
```

### 2-2.ndarray.resize(new_shape, refcheck=True)
Change shape and size of array in-place(그 자체의 값을 바꾸고 리턴을 하지않는다)

```python
a = np.arange(8)
b = a.resize((2,2))
print(f'original ndarray:\n{a}')
print(f'resized ndarray:\n{b}\n')

a = np.arange(8)
print(f'original ndarray:\n{a}')
a.resize((2,2))
print(f'resized ndarray:\n{a}')
```

## 3.ndarray.flatten and ndarray.ravel

### 3-1.ndarray.flatten(order='C')

```python
import numpy as np

M = np.arange(9)
N = M.reshape((3,-1))
O = N.flatten()

print(f'original ndarray:\n{M}')
print(f'resized ndarray:\n{N}')
print(f'flattened ndarray:\n{O}\n')

M = np.arange(27)
N = M.reshape((3,3,-1))
O = N.flatten()

print(f'original ndarray:\n{M}')
print(f'resized ndarray:\n{N}')
print(f'flattened ndarray:\n{O}')
```

### 3-2.ndarray.ravel([order])

```python
import numpy as np

M = np.arange(9)
N = M.reshape((3,-1))
O = N.ravel()

print(f'original ndarray:\n{M}')
print(f'resized ndarray:\n{N}')
print(f'flattened ndarray:\n{O}')
```

## 4.Memory Optimization

### 4-1.view of ndarrays

```python
import numpy as np

a = np.arange(5)
b = a.view()
print(f'b 수정전 a: {a}, b: {b}')
b[0] = 100
print(f'b 수정후 a: {a}, b: {b}\n')

a = np.arange(5)
b = a[0:3]
print(f'b 수정전 a: {a}, b: {b}')
b[...] = 10
print(f'b 수정후 a: {a}, b: {b}')
```

### 4-2.copy of ndarrays

```python
import numpy as np

a = np.arange(5)
b = a.copy()
print(f'b 수정전 a: {a}, b: {b}')
b[0] = 100
print(f'b 수정후 a: {a}, b: {b}\n')

a = np.arange(5)
b = a[0:3].copy()
print(f'b 수정전 a: {a}, b: {b}')
b[...] = 10
print(f'b 수정후 a: {a}, b: {b}\n')
```

### 4-3. base of ndarrays

```python
import numpy as np

a = np.arange(5)
b = a.copy()
c = a.view()
d = a[0:3]

print(b.base is a,'\n') # copy를 한것은 다른 메모리를 차지하기 때문에 원본 객체는 base가 아니다.

# --- view를 통해 만들어진 객체들의 base는 원본 객체이다.
print(c.base is a)
print(d.base is a)
# ---
```

### 4-4.APIs and copy, view

```python
# np.reshape

import numpy as np

a = np.arange(4)
b = np.reshape(a, (2,2))
print(f'b 수정전\na:\n{a}\nb:\n{b}\n')

print(f'b.base is a: {b.base is a}\n')

b[0][0] = 100
print(f'b 수정후\na:\n{a}\nb:\n{b}\n')
print('--------------------------------\n')

a = np.arange(4)
b = np.reshape(a, (2,2)).copy()
print(f'b 수정전\na:\n{a}\nb:\n{b}\n')

print(f'b.base is a: {b.base is a}\n')

b[0][0] = 100
print(f'b 수정후\na:\n{a}\nb:\n{b}')
```

```python
# np.resize

import numpy as np

a = np.arange(4)
b = np.resize(a, (2,2))
print(f'b 수정전\na:\n{a}\nb:\n{b}\n')

print(f'b.base is a: {b.base is a}\n')

b[0][0] = 100
print(f'b 수정후\na:\n{a}\nb:\n{b}\n')
```

```python
# ndarray.ravel

import numpy as np

a = np.random.randint(0,10,(2,3))
b = a.ravel()
print(f'b 수정전\na:\n{a}\nb:\n{b}\n')

print(f'b.base is a: {b.base is a}\n')

b[0] = 100

print(f'b 수정후\na:\n{a}\nb:\n{b}\n')
```

```python
# ndarray.flatten

import numpy as np

a = np.random.randint(0,10,(2,3))
b = a.flatten()
print(f'b 수정전\na:\n{a}\nb:\n{b}\n')

print(f'b.base is a: {b.base is a}\n')

b[0] = 100

print(f'b 수정후\na:\n{a}\nb:\n{b}\n')
```

## 5.ndarray.astype

### 5-1.ndarray.astpye

```python
import numpy as np

M = np.array([1,2,3], dtype = np.int8)
print(f'Original data type: {M.dtype}')
N = M.astype(np.float32)
print(f'Changed data type: {N.dtype}')
O = M.astype(np.uint32)
print(f'Changed data type: {O.dtype}')
```

```python
import numpy as np

M = np.random.uniform(-10,10, (3,3))
print(f'M =\n{M}, dtype = {M.dtype}\n')

N = M.astype(np.int32)
print(f'N =\n{N}, dtype = {N.dtype}\n')
```

### 5-2.bool dtype ndarrays 

```python
import numpy as np

bools = np.array([True, False])
print(f'bool: \n{bools}')

bools2ints = bools.astype(np.int32)
print(f'int: \n{bools2ints}')

bools2floats = bools.astype(np.float32)
print(f'float: \n{bools2floats}\n')

# -------------------------------------------- #

ints = np.array([-2, -1, 0, 1, 2])
ints2bools = ints.astype(np.bool)
print(f'int:\n{ints}')
print(f'bool:\n{ints2bools}\n')

floats = np.array([-2.5, -1.5, 0., 1.5, 2.5])
floats2bools = floats.astype(np.bool)
print(f'float:\n{floats}')
print(f'bool:\n{floats2bools}')
```
