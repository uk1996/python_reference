```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

import warnings
warnings.filterwarnings('ignore')
```

## Exercise 1

```python
np.random.seed(8)


n_data = 20
n_point = 200
scale = 3
x_mean_list = np.random.randint(-60, 60, 20)
y_mean_list = np.random.randint(-60, 60, 20)

cmap = cm.get_cmap('rainbow', lut=n_data)

fig, ax = plt.subplots(figsize=(15,15))
for data_idx in range(n_data):
    x_data = np.random.normal(loc=x_mean_list[data_idx],
    scale=scale,
    size=(n_point, ))
    
    y_data = np.random.normal(loc=y_mean_list[data_idx],
    scale=scale,
    size=(n_point, ))
    
    ax.scatter(x_data, y_data,
              color=cmap(data_idx),
              alpha=0.5)

x_ticks = np.arange(-60, 61, 20)
y_ticks = np.arange(-60, 61, 20)

ax.set_xticks(x_ticks)
ax.set_yticks(y_ticks)
    
    
ax.tick_params(axis='both',
              labelsize=20,
              length=8,
              width=1.5)

for spine_loc, spine in ax.spines.items():
    spine.set_linewidth(1.5)

    
    
plt.show()
```

## Exercise 3

```python
countries = ['Australia', 'Austria', 'Belgium', 'Canada', 'Chile',
             'Czech Republic', 'Denmark', 'Estonia', 'Finland', 'France',
             'Germany', 'Greece', 'Hungary', 'Iceland', 'Ireland',
             'Israel', 'Italy', 'Japan', 'Korea', 'Luxembourg',
             'Mexico', 'Netherlands', 'New Zealand', 'Norway', 'Poland',
             'Portuagl', 'Slovak Republic', 'Slovenia', 'Spain', 'Sweden',
             'Switzerland', 'Turkey', 'United Kingdom', 'United States']
population_denstiy = [3, 101, 367, 4, 23,
                     133, 130, 29, 16, 117,
                     227, 86, 106, 3, 65,
                     365, 203, 337, 501, 207,
                     60, 406, 17, 13, 122,
                     116, 110, 103, 91, 21,
                     194, 97, 257, 32]
private_expenditure = [1.3, 8.9, 10.0, 10.9, 12.8,
                    13.0, 13.2, 13.4, 13.5, 14.8,
                    15.0, 15.8, 16.7, 17.1, 17.2,
                    17.4, 18.1, 18.1, 18.7, 18.9,
                    19.0, 19.0, 19.5, 19.9, 20.0,
                    21.5, 21.6, 23.0, 24.0, 25.0,
                    25.9, 26.7, 28.0, 34.3]
gdp = [38.7, 37.4, 33.6, 37.5, 16.4,
                     24.5, 33.2, 19.3, 32.1, 32.0,
                     36.2, 19.8, 17.8, 37.7, 37.7,
                     29.4, 26.6, 32.0, 31.0, 67.9,
                     13.4, 38.4, 27.0, 48.2, 18.9,
                     20.9, 21.8, 24.2, 26.8, 36.2,
                     42.5, 13.9, 35.6, 45.7]

def data2size(data, min_size, max_size):
    data_copy = data.copy()
 
    min_val = np.min(data_copy)
    data_copy -= min_val
 
    max_val = np.max(data_copy)
    data_copy = data_copy / max_val
 
    interval = max_size - min_size
    data_copy *= interval
    data_copy += min_size
    return data_copy

gdp_size = data2size(gdp, 30, 80)

cmap = cm.get_cmap('tab20', lut=int(len(countries)/2))
c_arr = [cmap(i) for i in range(int(len(countries)/2))]

fig, ax = plt.subplots(figsize=(12, 8))
# group1 scatter plotting
ax.scatter(population_denstiy[:17], private_expenditure[:17],
          s=gdp_size[:17]**2,
          c=c_arr,
          alpha=0.3,
          hatch='/')
# group2 scatter plotting
ax.scatter(population_denstiy[17:], private_expenditure[17:],
          s=gdp_size[17:]**2,
          c=c_arr,
          alpha=0.3,
          hatch='.')

ax.set_ylim([-3, 40])

ax.set_xlabel('population_denstiy(Inh./km2)', fontsize=15)
ax.set_ylabel('private_expenditure', fontsize=15)

ax.grid(linestyle='--')

# conty legend
for country_idx in range(17):
    ax.scatter([], [],
              s=500,
               facecolor=c_arr[country_idx],
               hatch='/',
               alpha=0.3,
              label=countries[country_idx]
              )
    
for country_idx in range(17, 34):
    ax.scatter([], [],
              s=500,
               facecolor=c_arr[country_idx%17],
               hatch='.',
               alpha=0.3,
              label=countries[country_idx]
              )
    
ax.legend(ncol=2,
          bbox_to_anchor=(1, 1), loc='upper left',
         labelspacing=1.5)

# size legend
legend_gdp = [i for i in range(10, 61, 15)]
legend_sizes = data2size(legend_gdp, 30, 80)

ax2 = ax.twinx()
ax2.get_xaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)

for legend_idx, legend_size in enumerate(legend_sizes):
    ax2.scatter([], [],
               color='tab:blue',
               s=legend_sizes[legend_idx]**2,
               label=legend_gdp[legend_idx])
ax2.legend(bbox_to_anchor=(0.5, 1.05), loc='lower center',
          title='GDP Value',
          title_fontsize=30,
          ncol=len(legend_sizes),
           fontsize=20,
          columnspacing=2,
          labelspacing=2,
           edgecolor='None',
          handletextpad=1.4)

fig.subplots_adjust(right=0.8, left=0.05)

plt.show()
```
