```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm

import warnings
warnings.filterwarnings('ignore')
```

## Exercise 1

```python
# rcParams setting
plt.style.use('seaborn')
plt.rcParams['lines.linewidth'] = 2

# data setting
msr_x = np.linspace(0, 2E5, 10)
msr_y = np.linspace(0, 8E5, 10)

vatex_x = np.linspace(0, 4.1E5, 20)
vatex_en_y = np.linspace(0, 29E5, 20)
vatex_zh_y = np.linspace(0, 28E5, 20)

# plotting
fig, ax = plt.subplots(figsize=(10, 6.5))
ax.plot(vatex_x, vatex_en_y,
 color='g',
 label='VATEX-en')
ax.plot(vatex_x, vatex_zh_y,
 linestyle='--',
 color='b',
 label='VATEX-zh') 
ax.plot(msr_x, msr_y,
 linestyle='-.',
 color='r',
 label='MSR-VTT')

# axis customizing
ax.set_ylim([-1E5, 3E6])
y_ticks = np.linspace(0, 25E5, 6)
ax.set_yticks(y_ticks)
 
# tick & grid customizing
ax.ticklabel_format(axis='both', style='plain')
ax.tick_params(labelsize=15)
ax.grid(linewidth=2)

# legend customizing
ax.legend(loc='lower right', fontsize=15)

# label customizing
ax.set_xlabel('number of captions',
 fontsize=15)
ax.set_ylabel('number of types',
 fontsize=15)

fig.tight_layout()
plt.show()
```

## Exercise 2

```python
# rcParams setting
plt.style.use('default')
plt.rcParams['font.family'] = 'serif'

# data setting
data_dict_list = []
data_dict1 = {'DF':[90, 99, 100, 100, 100],
 'F2F':[77, 98, 99, 100, 100],
 'FS':[91,96,98,99,100],
 'NT':[94,95,97,98,100],
 'All':[93,94,96,97,100]}
data_dict_list.append(data_dict1)
data_dict2 = {'DF':[77, 91, 95, 98, 99],
 'F2F':[60, 91, 95, 98, 99],
 'FS':[81, 91, 95, 98, 99],
 'NT':[77, 85, 90, 92, 95],
 'All':[68, 82, 88, 92, 96]}
data_dict_list.append(data_dict2)
data_dict3 = {'DF':[66, 76, 83, 91, 94],
 'F2F':[55, 67, 72, 88, 91],
 'FS':[60, 75, 80, 88, 91],
 'NT':[66, 71, 72, 80, 82],
 'All':[60, 66, 67, 77, 81]}
data_dict_list.append(data_dict3)
line_name_list = ['Deepfakes', 'Face2Face', 'FaceSwap',
 'NeuralTextures', 'All']
ylabel_list = ['Accuracy Raw', 'Accuracy HQ', 'Accuracy LQ']



fig, axes = plt.subplots(3, 1, figsize=(7, 9))

x_ticks = range(5)
xtick_labels = [10, 50, 100, 300, 1000]
y_ticks = range(50, 101, 10)

color_list = ['royalblue', 'darkorange','gray', 'cornflowerblue', 'gold']

for ax_idx, ax in enumerate(axes):
    ax.set_xticks(x_ticks)
    ax.set_xticklabels(xtick_labels)
    ax.set_ylim([49, 101])
    ax.set_yticks(y_ticks)
    ax.grid(axis='y')
    ax.tick_params(left=False, bottom=False)
    ax.set_ylabel(ylabel_list[ax_idx], fontsize=12)
    
    for spine_loc, spine in ax.spines.items():
        spine.set_visible(False)
    
    data = data_dict_list[ax_idx]
    for line_idx, line_value in enumerate(data.values()):
        ax.plot(x_ticks, line_value, color=color_list[line_idx], linewidth=2.5, label=line_name_list[line_idx])
    
    if ax_idx == len(data_dict_list) - 1:
        ax.legend(bbox_to_anchor=(0.5, -0.3), loc='upper center', ncol=len(line_name_list), fontsize=8, edgecolor='white')
        ax.set_xlabel('Number of videos', fontsize=12)

fig.subplots_adjust(hspace=0.3)
plt.show()
```

## Exercise 4

```python
# data setting
name_list = ['PL', 'NL', 'NL-<SeINL', 'NL->SeINL->SeIPL']
x_data = np.linspace(0.5, 1, 30)
y_data = np.empty(shape = (0,len(x_data)))
gamma_list = [6,7, 8, 12]
for i in range(len(gamma_list)):
    y = (-0.5)*np.power(x_data, gamma_list[i]) + 1
    y = y.reshape(1,len(x_data))
    y_data = np.vstack((y_data, y))
    

fig, ax = plt.subplots(figsize=(5, 5))

ax.set_xlim([0.5, 1])
ax.set_ylim([0.5, 1])
ax.grid('-', linewidth=1.2, alpha=0.5)

ax.set_xlabel('recall')
ax.set_ylabel('precision')

for spine_loc, spine in ax.spines.items():
    spine.set_linewidth(2)
    spine.set_alpha(0.5)

for y_idx, y in enumerate(y_data):
    ax.plot(x_data, y, marker='o', markersize=5, label=name_list[y_idx])

ax.legend(fontsize=8.2)

fig.tight_layout()
plt.show()
```

## Exercise 5

```python
# data setting
name_list = ['filter/natural test', 'column/natural test', 'irregular/natural test', 
             'filter/adversarial test', 'column/adversarial test', 'irregular/adversarial test']
x_loc = [2**i for i in range(5)]
data_dict = {'FN':[50, 65, 70, 75, 76],
 'CN':[61, 68, 70, 74, 76],
 'IN':[76, 78, 77, 78, 76],
 'FA':[34, 38, 41, 42, 43],
 'CA':[36, 40, 42, 43, 43],
 'IA':[42, 41, 44, 44, 43]}

color_list = ['r', 'g', 'c']
line_style_list = ['-', '--']
marker_style_list = ['v','^', 's', 'o', '+',  'x' ]

xtick_labels = [1, 2, 4, 8, 16]

fig, ax = plt.subplots(figsize=(6, 4))

ax.set_xscale('log', base=2)

ax.set_xlabel('w', fontsize=13)
ax.set_ylabel('Accuracy(%)', fontsize=13)

ax.set_xticks(x_loc)
ax.set_xticklabels(xtick_labels)

for spine_loc, spine in ax.spines.items():
    spine.set_linewidth(1.5)

for idx, (label, data) in enumerate(data_dict.items()):
    c_idx = idx % 3
    l_idx = idx // 3
    ax.plot(x_loc, data, color=color_list[c_idx], linestyle=line_style_list[l_idx], 
            marker=marker_style_list[idx], label=name_list[idx])

ax.set_ylim([0, 100])
    
ax.legend(bbox_to_anchor=(0.5, 0), loc='lower center', ncol=2)
    
fig.tight_layout()
plt.show()
```

## Exercise 6

```python
# data setting
data = [2, 6, 7, 15, 5.5, 8.5, 8, 12, 20, 15, 0.2]

xticks = np.arange(len(data)) + 1
major_yticks = [1, 10]
#minor_yticks = [j*10**i for i in range(-1, 2) for j in range(1, 10)]

max_value = np.array(data).max()
max_value_idx = np.array(data).argmax() + 1
min_value = np.array(data).min()
min_value_idx = np.array(data).argmin() + 1

fig, ax = plt.subplots(figsize=(6, 5))

ax.set_yscale('log', base=10)
ax.set_ylim([0.1, 50])

ax.set_xlim([0, 12])
xlim = ax.get_xlim()

ax.set_xticks(xticks)
ax.set_yticks(major_yticks)
#ax.set_yticks(minor_yticks, minor=True)
ax.tick_params(width=1.5, direction='in')
ax.tick_params(which='minor', direction='in')

ax.set_xlabel('Blocks->')
ax.set_ylabel('Top Hessian Eigenvalue->')

ax.grid(which='both')

for spine_loc, spine in ax.spines.items():
    spine.set_linewidth(1.2)
    spine.set_alpha(0.8)

ax.plot(xticks, data, color='black', marker='o', label='ResNet20 On Cifar-10')
ax.plot([max_value_idx, xlim[1]], [max_value, max_value], color='gray', linestyle='--')
ax.plot([min_value_idx, xlim[1]], [min_value, min_value], color='gray', linestyle='--')


ax.legend(loc='upper right', edgecolor='white', facecolor='white')


fig.tight_layout()
plt.show()
```
