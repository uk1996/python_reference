```python
import matplotlib.pyplot as plt
import numpy as np
```

## Scatter Plot Basic

### ax.plot and ax.sactter

```python
n_data = 100
x_data = np.random.normal(0, 1, (n_data, ))
y_data = np.random.normal(0, 1, (n_data, ))

fig, axes = plt.subplots(1, 2, figsize=(10, 5))

axes[0].scatter(x_data, y_data, 
                s=100, # 넓이의 개념
                color='red')
axes[1].plot(x_data, y_data, 'o', 
             markersize=10, # 길이의 개념
             color='red')

plt.show()
```

```python
x_min, x_max = -5, 5
n_data = 300

x_data = np.random.uniform(x_min, x_max, n_data)
y_data = x_data + 0.5*np.random.normal(0, 1, (n_data, ))

pred_x = np.linspace(x_min, x_max, n_data)
pred_y = pred_x

fig, ax = plt.subplots(figsize=(5, 5))

ax.scatter(x_data, y_data)
ax.plot(pred_x, pred_y, color='red', linewidth=3)

plt.show()
```

### Size Array and Color Array

```python
n_data = 10
x_data = np.linspace(0, 10, n_data)
y_data = np.linspace(0, 10, n_data)

s_arr = np.linspace(10, 500, n_data) # 3D information
c_arr = [(c/10, c/10, c/10) for c in range(n_data)]

fig, ax = plt.subplots(figsize=(8, 8))

ax.scatter(x_data, y_data,
          s=s_arr,
          c=c_arr)

plt.show()
```

```python
n_data = 500
x_data = np.random.normal(0, 1, size=(n_data, ))
y_data = np.random.normal(0, 1, size=(n_data, ))
s_arr = np.random.uniform(100, 500, size=(n_data))
c_arr = [np.random.uniform(0, 1, 3) for _ in range(n_data)]

fig, ax = plt.subplots(figsize=(8, 8))

ax.scatter(x_data, y_data,
          s=s_arr,
          c=c_arr,
          alpha=0.3)

plt.show()
```

```python
def data2size(data, min_size, max_size):
    data_copy = data.copy()
    
    min_val = np.min(data_copy)
    data_copy -= min_val
    
    max_val = np.max(data_copy)
    data_copy = data_copy/max_val
    
    interval = max_size - min_size
    data_copy *= interval
    data_copy += min_size
    
    return data_copy

n_data = 10
x_data = np.linspace(0, 10, n_data)
y_data = np.linspace(0, 10, n_data)

z_data = np.random.normal(0, 1, n_data)
s_arr = data2size(z_data, 100, 2000)

fig, ax = plt.subplots(figsize=(10, 10))

ax.scatter(x_data, y_data,
          s=s_arr)

plt.show()
```

### Color Array at c Argument

```python
import matplotlib.cm as cm
```

```python
PI = np.pi
n_point = 100
t = np.linspace(-4*PI, 4*PI, n_point)
sin = np.sin(t)

cmap = cm.get_cmap('Reds', lut=n_point)
c_arr = [cmap(c_idx) for c_idx in range(n_point)]

fig, ax = plt.subplots(figsize=(10, 7))

ax.scatter(t, sin, 
          s=100,
          c=c_arr)

plt.show()
```

## Advanced Markers and Legend

### Advanced Markers

```python
fig, ax = plt.subplots(figsize=(5, 5))

ax.scatter(0,0,
          color='r')
ax.scatter(0, 0,
          s=10000,
          facecolor='None',
          edgecolor='black',
          linewidth=5)

plt.show()
```

```python
n_data = 200
x_data = np.random.normal(0, 1, (n_data, ))
y_data = np.random.normal(0, 1, (n_data, ))

fig, ax = plt.subplots(figsize=(7, 7))

ax.scatter(x_data, y_data,
          s=100,
          facecolor='None',
          edgecolor='black',
          linewidth=3,
          alpha=0.5)

plt.show()
```

```python
n_class = 5
n_data = 30
center_pt = np.random.uniform(-20, 20, (n_class, 2))

cmap = cm.get_cmap('tab10')
colors = [cmap(i) for i in range(n_class)]

data_dict = {'class' + str(i):None for i in range(n_class)}
for class_idx in range(n_class):
    center = center_pt[class_idx]
    x_data = center[0] + 2*np.random.normal(0, 1, (1, n_data))
    y_data = center[1] + 2*np.random.normal(0, 1, (1, n_data))
    data = np.vstack((x_data, y_data))
    data_dict['class' + str(class_idx)] = data

fig, ax = plt.subplots(figsize=(10, 10))

for class_idx in range(n_class):
    data = data_dict['class' + str(class_idx)]
    ax.scatter(data[0], data[1],
              s=1000,
              facecolor='None',
              edgecolor=colors[class_idx],
              linewidth=2,
              alpha=0.5,
              label='class' + str(class_idx))

ax.legend()
    
plt.show()
```

### Advanced Legend

```python
np.random.seed(0)

n_class = 10
n_data = 30
center_pt = np.random.uniform(-20, 20, (n_class, 2))

cmap = cm.get_cmap('tab10')
colors = [cmap(i) for i in range(n_class)]

data_dict = {'class' + str(i):None for i in range(n_class)}
for class_idx in range(n_class):
    center = center_pt[class_idx]
    x_data = center[0] + 2*np.random.normal(0, 1, (1, n_data))
    y_data = center[1] + 2*np.random.normal(0, 1, (1, n_data))
    data = np.vstack((x_data, y_data))
    data_dict['class' + str(class_idx)] = data

fig, ax = plt.subplots(figsize=(10, 10))

for class_idx in range(n_class):
    data = data_dict['class' + str(class_idx)]
    ax.scatter(data[0], data[1],
              s=1000,
              facecolor='None',
              edgecolor=colors[class_idx],
              linewidth=2,
              alpha=0.5,
              label='class' + str(class_idx))

ax.legend(bbox_to_anchor=(0.5, 0), loc='upper center',
         ncol=5,
         fontsize=15,
         title='Clasees',
         title_fontsize=20,
         facecolor='None',
         edgecolor='None',
         labelspacing=1.4, # rowspacing
         columnspacing=1.8)
    
plt.show()
```

## Size Legend

### Fake Legend

```python
c_arr = ['r', 'g', 'b']
s_arr = [500, 2000, 5000]

x_loc = [-2, 0, 2]
y_loc = x_loc

fig, ax = plt.subplots(figsize=(10, 10))

for plot_idx in range(len(c_arr)):
    ax.scatter(x_loc[plot_idx], y_loc[plot_idx],
              c=c_arr[plot_idx],
              s=s_arr[plot_idx],
              label=' data '+str(plot_idx))

ax.legend(bbox_to_anchor=(1.05, 0.5), loc='center left',
         labelspacing=10,
         fontsize=20,
         edgecolor='None')
    
plt.show()
```

### Size Legend

```python
legend_sizes = np.array([10, 20, 30, 40, 50])

fig, ax = plt.subplots(figsize=(10, 10))

for legend_size in legend_sizes:
    ax.scatter([], [],
              facecolor='None',
              edgecolor='blue',
               linewidth=2,
              s=legend_size**2,
              label=str(legend_size))
    
ax.legend(bbox_to_anchor=(1, 0.5), loc='center left',
         title='Point Size',
         title_fontsize=30,
         fontsize=25,
         labelspacing=1.2)

plt.show()
```

### Normal Legend with Size Legend

```python
np.random.seed(0)

n_class = 5
n_data = 100

cmap = cm.get_cmap('tab20')
colors = [cmap(i) for i in range(n_class)]

center_pt = np.random.uniform(-10, 10, (n_class, 2))

fig, ax = plt.subplots(figsize=(10, 10))

for class_idx in range(n_class):
    center = center_pt[class_idx]
    x_data = center[0] + np.random.normal(0, 1, (n_data,))
    y_data = center[1] + np.random.normal(0, 1, (n_data,))
    s_arr = np.random.uniform(10, 40, (n_data,))
    
    ax.scatter(x_data, y_data,
              s=s_arr**2,
              facecolor='None',
              edgecolor=colors[class_idx],
              label='class ' + str(class_idx))

ax.legend(loc='upper right',
         fontsize=25)

ax2 = ax.twinx()
ax2.get_xaxis().set_visible(False)
ax2.get_yaxis().set_visible(False)

legend_sizes = [10, 20, 30, 40]
for legend_size in legend_sizes:
    ax2.scatter([], [],
               facecolor='None',
               edgecolor='black',
               linewidth=3,
               s=legend_size**2,
               label=str(legend_size))
ax2.legend(bbox_to_anchor=(1, 0.5),
          loc='center left',
          title='Point Size',
          title_fontsize=30,
          fontsize=25,
          labelspacing=1.2)

plt.show()
```
