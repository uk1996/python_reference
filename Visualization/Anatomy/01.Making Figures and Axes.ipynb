```python
# 모듈 임포트
import matplotlib.pyplot as plt
import numpy as np
```

## plt.figure, fig.add_subplot

```python
# Figure 만들기
fig = plt.figure(figsize=(7,7),       # figure 크기
                 facecolor='linen')   # figure 배경색(named color)

# Axes 만들기
ax1 = fig.add_subplot(2,3,1)
ax2 = fig.add_subplot(2,3,2)
ax3 = fig.add_subplot(2,3,4)
ax4 = fig.add_subplot(2,3,5)
ax5 = fig.add_subplot(1,3,3)
```

## plt.subplots(Making Figure and Axes simultaneously)

```python
fig,axes = plt.subplots(2,1,
                        figsize = (7,7),
                        facecolor = 'linen')
```

### Unpacking

```python
fig , (ax1,ax2) = plt.subplots(2,1,
                        figsize = (7,7),
                        facecolor = 'linen')
```

### Indexing

#### 1-D Axes

```python
fig, axes = plt.subplots(3,1,
                        figsize = (7,7),
                        facecolor = 'linen')

for idx in range(3):
    print(axes[idx])
```

#### 2-D Axes

```python
fig, axes = plt.subplots(2,2,
                        figsize = (7,7),
                        facecolor = 'linen')

print('axes shape: ',axes.shape)

for row_idx in range(2):
    for col_idx in range(2):
        print(f'axes[{row_idx},{col_idx}]: {axes[row_idx,col_idx]}')
```

```python
# ndarray.flat and enumerate
fig, axes = plt.subplots(2,2,
                        figsize = (7,7),
                        facecolor = 'linen')

print('axes shape: ',axes.shape)

for ax_idx,ax in enumerate(axes.flat): # matrix의 row들을 연결하여 하나의 list로 만들어준다
    print(ax_idx,ax)
```

## plt.subplot2grid(More Complex Arrangement)

```python
fig = plt.figure(figsize=(7,7),facecolor='linen')

ax1 = plt.subplot2grid((3,3),(0,0),
                       colspan = 3,
                       fig=fig)
ax2 = plt.subplot2grid((3,3),(1,0),
                       colspan = 2,
                       fig=fig)
ax3 = plt.subplot2grid((3,3),(1,2),
                       rowspan = 2,
                       fig=fig)

ax4 = plt.subplot2grid((3,3),(2,0),
                       fig=fig)
ax5 = plt.subplot2grid((3,3),(2,1),
                       fig=fig)
```

### Practice

```python
fig = plt.figure(figsize=(7,7),facecolor='linen')
axes = np.empty(shape=(0,))

axes = np.append(axes,plt.subplot2grid((5,4),(0,0),rowspan = 2,colspan = 2,fig=fig))
axes = np.append(axes,plt.subplot2grid((5,4),(0,2),rowspan = 2,colspan = 2,fig=fig))

for row_idx in range(2,5):
    for col_idx in range(4):
        axes = np.append(axes,plt.subplot2grid((5,4),(row_idx,col_idx),fig=fig))
```

## fig.add_axes(Arbitrary Locations and Sizes of Axes)

```python
'''
fig = plt.figure()
ax = fig.add_axes([left,bottom,width,height])
'''

left,bottom = 0.1,0.1
spacing = 0.05
width1,height1 = 0.5,0.5

width2 = 1-(2*left+width1+spacing)
height2 = 1-(2*bottom+height1+spacing)

rect1 = [left,bottom,width1,height1]
rect2 = [left,bottom+height1+spacing,1-2*left,height2]
rect3 = [left+width1+spacing,bottom,width2,height1]

fig = plt.figure(figsize=(7,7),facecolor='linen')

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2)
ax3 = fig.add_axes(rect3)
```
