```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1.ax.set_title, ax.set_xlabel, ax.set_ylabel

```python
fig,ax = plt.subplots(figsize = (7,7))
ax.set_title('Title',fontsize = 20)
ax.set_xlabel('X Label',fontsize = 15)
ax.set_ylabel('Y Label',fontsize = 15)
```

```python
title_list = ['Ax' + str(i) for i in range(4)]
xlabel_list = ['X Label' + str(i) for i in range(4)]
ylabel_list = ['Y Label' + str(i) for i in range(4)]

print('title_list:',title_list)
print('xlabel_list:',xlabel_list)
print('ylabel_list:',ylabel_list)

fig,axes = plt.subplots(2,2,figsize = (7,7))

for ax_idx,ax in enumerate(axes.flat):
    ax.set_title(title_list[ax_idx],fontsize=30)
    ax.set_xlabel(xlabel_list[ax_idx],fontsize=20)
    ax.set_ylabel(ylabel_list[ax_idx],fontsize=20)

print('\n')
print('figure를 보면 title,label 들이 겹치는 것을 볼수 있다. 그러므로 Custimizing하여 겹치지 않게 해주어야한다.') 
```

## 2.fig.tight_layout(Axes Layout Adjustment)

```python
title_list = ['Ax' + str(i) for i in range(4)]
xlabel_list = ['X Label' + str(i) for i in range(4)]
ylabel_list = ['Y Label' + str(i) for i in range(4)]

print('title_list:',title_list)
print('xlabel_list:',xlabel_list)
print('ylabel_list:',ylabel_list)
print('\n')

fig1,axes1 = plt.subplots(2,2,figsize = (7,7),facecolor='linen')
fig2,axes2 = plt.subplots(2,2,figsize = (7,7),facecolor='linen')
fig3,axes3 = plt.subplots(2,2,figsize = (7,7),facecolor='linen')
fig4,axes4 = plt.subplots(2,2,figsize = (7,7),facecolor='linen')


for ax_idx,ax in enumerate(axes1.flat):
    ax.set_title(title_list[ax_idx],fontsize=30)
    ax.set_xlabel(xlabel_list[ax_idx],fontsize=20)
    ax.set_ylabel(ylabel_list[ax_idx],fontsize=20)
for ax_idx,ax in enumerate(axes2.flat):
    ax.set_title(title_list[ax_idx],fontsize=30)
    ax.set_xlabel(xlabel_list[ax_idx],fontsize=20)
    ax.set_ylabel(ylabel_list[ax_idx],fontsize=20)
for ax_idx,ax in enumerate(axes3.flat):
    ax.set_title(title_list[ax_idx],fontsize=30)
    ax.set_xlabel(xlabel_list[ax_idx],fontsize=20)
    ax.set_ylabel(ylabel_list[ax_idx],fontsize=20)
for ax_idx,ax in enumerate(axes4.flat):
    ax.set_title(title_list[ax_idx],fontsize=30)
    ax.set_xlabel(xlabel_list[ax_idx],fontsize=20)
    ax.set_ylabel(ylabel_list[ax_idx],fontsize=20)



fig1.tight_layout()             # pad = default
fig2.tight_layout(pad = 0.3)    # pad = 0.3
fig3.tight_layout(pad = 3)      # pad = 3
fig4.tight_layout(pad = 7)      # pad = 7
```

## 3.fig.subplots_adjust(More Customized Layout)

### 3-1. Usage

```python
# Positon Arguments
fig,axes = plt.subplots(2,2,figsize=(7,7),facecolor='linen')

# --- tick 제거
for ax_idx,ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
# ---

fig.subplots_adjust(bottom=0,top=0.9,left=0.05,right=0.7) # bottm,top,left,right 간격 조정
```

```python
# Spacing Arguments
fig,axes = plt.subplots(3,3,figsize=(7,7),facecolor='linen')

# --- tick 제거
for ax_idx,ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
# ---

fig.subplots_adjust(wspace=0,hspace=0.2)
```

### 3-2.Practice

```python
fig = plt.figure(figsize=(7,7),facecolor='linen')
axes = np.empty(shape=(0,7))

# --- Figure,Axes 만들기
for g_idx in range(2):
    axes_g = np.empty(shape=(0,))
    main = plt.subplot2grid((5,4),(0,2*g_idx),
                            rowspan=2,colspan=2,
                            fig=fig)
    axes_g = np.append(axes_g,main)

    for row_idx in range(2,5):
        for col_idx in range(2):
            ax = plt.subplot2grid((5,4),(row_idx,col_idx+2*g_idx),
                                  fig=fig)
            axes_g = np.append(axes_g,ax)
    axes_g = axes_g.reshape(1,-1)
    axes = np.vstack((axes,axes_g))
print('axes shape:',axes.shape)
# ---

# --- tick 제거
for ax_idx,ax in enumerate(axes.flat):
    ax.get_xaxis().set_visible(False)
    ax.get_yaxis().set_visible(False)
# ---

# --- fig.subplots_adjust(간격조정)
fig.subplots_adjust(bottom=0.03,top=0.97,
                    left=0.03,right=0.97,
                    hspace=0,wspace=0)
# ---
```

```python
fig = plt.figure(figsize=(7,7))

ax1_1 = fig.add_subplot(2,2,1)
ax1_2 = fig.add_subplot(2,2,2,sharey=ax1_1)
ax2_1 = fig.add_subplot(2,2,3,sharex=ax1_1)
ax2_2 = fig.add_subplot(2,2,4,sharex=ax1_2,sharey=ax2_1)

ax1_1.set_ylim([0,10])
ax2_1.set_ylim([0,5])
ax2_1.set_xlim([0,100])
ax2_2.set_xlim([0,500])

fig.tight_layout()
```

```python
fig = plt.figure(figsize=(7,7))
axes = np.empty(shape=(0,))

for ax_idx in range(1,1+3):
    if ax_idx == 1:
        axes = np.append(axes,fig.add_subplot(3,1,ax_idx))
    else:
        axes = np.append(axes,fig.add_subplot(3,1,ax_idx,sharex=axes[0]))

axes[2].set_xlim([10,1000])

fig.tight_layout()
```

```python
fig = plt.figure(figsize=(7,7))
n_row,n_col = 3,3
axes = np.empty(shape=(0,3))

for r_idx in range(n_row):
    axes_row = np.empty(shape=(0,))
    if r_idx == 0:
        for c_idx in range(n_col):
            if c_idx == 0:
                ax = fig.add_subplot(n_row,n_col,r_idx*n_col+c_idx+1)
            else:
                ax = fig.add_subplot(n_row,n_col,r_idx*n_col+c_idx+1,sharey=axes_row[0])
        
            axes_row = np.append(axes_row,ax)
        axes_row = axes_row.reshape(1,-1)
    else:
        for c_idx in range(n_col):
            if c_idx == 0:
                ax = fig.add_subplot(n_row,n_col,r_idx*n_col+c_idx+1,
                                     sharex=axes[0,c_idx])
            else:
                ax = fig.add_subplot(n_row,n_col,r_idx*n_col+c_idx+1,
                                     sharex=axes[0,c_idx],
                                     sharey=axes_row[0])
            axes_row = np.append(axes_row,ax)
        axes_row = axes_row.reshape(1,-1)
    axes = np.vstack((axes,axes_row))

print('axes shape:',axes.shape)

axes[0,0].set_ylim([0,100])
axes[1,0].set_ylim([0,200])
axes[2,0].set_ylim([0,300])
axes[2,0].set_xlim([0,10])
axes[2,1].set_xlim([0,20])
axes[2,2].set_xlim([0,30])

fig.tight_layout()
```

### 4-4.Using fig.add_axes

```python
left,bottom = 0.05,0.05
spacing = 0.05
width1,height1 = 0.6,0.6

width2 = 1 - (2*left+width1+spacing)
height2 = 1 - (2*bottom+height1+spacing)

rect1 = [left,bottom,width1,height1]
rect2 = [left,bottom+height1+spacing,width1,height2]
rect3 = [left+width1+spacing,bottom,width2,height1]

fig = plt.figure(figsize=(7,7),facecolor='linen')

ax1 = fig.add_axes(rect1)
ax2 = fig.add_axes(rect2,sharex=ax1)
ax3 = fig.add_axes(rect3,sharey=ax1)

ax1.set_xlim([0,100])
ax1.set_ylim([0,100])

fig.tight_layout()
```

### 4-5.Practice

```python
fig = plt.figure(figsize=(7,7))

n_row,n_col = 3,4
axes = np.empty(shape=(0,n_col))

xlabel_list = ['Iteration 01','Iteration 11','Iteration 21','Iteration 31']
ylabel_list = ['Log-Likelihood','Negative Grad.','Probability']

for r_idx in range(n_row):
    axes_row = np.empty(shape=(0,))
    if r_idx == 0:
        for c_idx in range(n_col):
            if c_idx == 0:
                ax = fig.add_subplot(n_row,n_col,r_idx*n_col+1+c_idx)
            else:
                ax = fig.add_subplot(n_row,n_col,r_idx*n_col+1+c_idx,sharey=axes_row[0])
            axes_row = np.append(axes_row,ax)
        axes_row = axes_row.reshape(1,-1)
    else:
        for c_idx in range(n_col):
            if c_idx == 0:
                ax = fig.add_subplot(n_row,n_col,r_idx*n_col+1+c_idx,
                                     sharex=axes[0,c_idx])
            else:
                ax = fig.add_subplot(n_row,n_col,r_idx*n_col+1+c_idx,
                                     sharex=axes[0,c_idx],
                                     sharey=axes_row[0])
            axes_row = np.append(axes_row,ax)
        axes_row = axes_row.reshape(1,-1)
    
    axes = np.vstack((axes,axes_row))

print('axes shape:',axes.shape)

axes[0,0].set_ylim([4,6])
axes[1,0].set_ylim([2,4])
axes[2,0].set_ylim([0,2])
axes[2,0].set_xlim([0,100])
axes[2,1].set_xlim([0,200])
axes[2,2].set_xlim([0,300])
axes[2,3].set_xlim([0,400])

for ax_idx,ax in enumerate(axes.flat):
    if ax_idx % n_col == 0:
        ax.set_ylabel(ylabel_list[ax_idx // n_col])
    
    if ax_idx >= 2*n_col:
        ax.set_xlabel(xlabel_list[ax_idx - 2*n_col])
    
    if ax_idx % n_col != 0:
        ax.get_yaxis().set_visible(False)
    
    if ax_idx < 2*n_col:
        ax.get_xaxis().set_visible(False)

fig.tight_layout()
fig.subplots_adjust(wspace=0.01,hspace=0.2)
```

## 5.Twinx and Axis Scale

### 5-1.ax.twinx(Different Y Values)

```python
fig = plt.figure(figsize=(7,7))

ax1 = fig.add_subplot()
ax2 = ax1.twinx()

ax1.set_xlim([0,100])
ax1.set_ylim([-1,1])
ax2.set_ylim([0,100])

ax1.set_title('Twinx Graph',fontsize=30)
ax1.set_ylabel('Data 1',fontsize=20)
ax2.set_ylabel('Data 2',fontsize=20)

fig.tight_layout()
```

### 5-2.ax.set_yscale(Axis scale)

```python
fig,axes = plt.subplots(2,2,
                        figsize=(7,7))

axes[0,1].set_yscale('log') # log scale
axes[1,0].set_yscale('log',basey=2) # 밑이 2인 log scale
axes[1,1].set_yscale('logit') # logit scale

fig.tight_layout()
```
