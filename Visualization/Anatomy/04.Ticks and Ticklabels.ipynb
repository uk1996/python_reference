```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1.ax.tick_params

### 1-1. label, lenght, width Arguments

```python
fig,axes = plt.subplots(1,2,figsize=(14,7))


axes[1].tick_params(labelsize=20,    # label 크기
               length = 10,          # tick 길이
               width = 3)            # tick 굵기
```

### 1-2.Tick Locations

```python
fig,axes = plt.subplots(2,2,figsize=(14,14))


axes[0,0].tick_params(labelsize=20,        
               length = 20,         
               width = 5,           
               bottom=False,
               labelbottom=False)

axes[0,1].tick_params(labelsize=20,        
               length = 10,         
               width = 3,           
               left=False,
               labelleft=False)

axes[1,0].tick_params(labelsize=20,        
               length = 10,         
               width = 3,           
               bottom=False, labelbottom=False,
               top=True, labeltop=True)

axes[1,1].tick_params(labelsize=20,        
               length = 10,         
               width = 3,           
               bottom=False, labelbottom=False,
               left=False, labelleft=False,
               top=True, labeltop=True,
               right=True,labelright=True)
```

### 1-3.rotation Argument

```python
fig,axes = plt.subplots(1,3,figsize=(21,7))

axes[0].tick_params(labelsize=20,
                    length=10,
                    width=3)

axes[1].tick_params(labelsize=20,
                    length=10,
                    width=3,
                    rotation=30)

axes[2].tick_params(labelsize=20,
                    length=10,
                    width=3,
                    rotation=-30)
```

### 1-4.x,y Axis Ticks

```python
fig,axes = plt.subplots(1,3,figsize=(21,7))

axes[0].tick_params(labelsize=20,
                    length=10,
                    width=3,
                    rotation=30)

axes[1].tick_params(axis='x',
                    labelsize=20,
                    length=10,
                    width=3,
                    rotation=30)

axes[2].tick_params(axis='y',
                    labelsize=20,
                    length=10,
                    width=3,
                    rotation=30)
```

### 1-5.Practice

## 2.ax.set_xticks

### 2-1.labelsize Argument

```python
fig,axes = plt.subplots(2,2,figsize=(14,14))

axes[0,0].tick_params(labelsize=20)

xticks = [i for i in range(10)]
axes[0,1].set_xticks(xticks)
axes[0,1].tick_params(labelsize=20)

axes[1,0].set_xlim([0,10])
axes[1,0].tick_params(labelsize=20)

axes[1,1].set_xticks([0,1,5,10])
axes[1,1].tick_params(labelsize=20)
```

```python
# Regular Intervals
fig,ax = plt.subplots(figsize=(14,7))

xticks = [i for i in range(0,101,20)]
ax.set_xticks(xticks)

ax.tick_params(axis='x',
               labelsize=20,
               length=10,
               width=3,
               rotation=30)
```

### 2-2.Major and Minor Ticks

```python
fig,ax = plt.subplots(figsize=(14,7))

major_xticks = [i for i in range(0,101,20)]
minor_xticks = [i for i in range(0,101,5)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks,
              minor=True)

ax.tick_params(axis='x',
               labelsize=20,
               length=10,
               width=3,
               rotation=30)

ax.tick_params(axis='x',
               which='minor',
               length=5,
               width=2)
```

### 2-3.x,y Ticks

```python
fig,ax = plt.subplots(figsize=(14,7))

major_xticks = [i for i in range(0,101,20)]
minor_xticks = [i for i in range(0,101,5)]
major_yticks = [i for i in range(0,11,2)]
minor_yticks = [i for i in range(0,11)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks,
              minor=True)

ax.tick_params(axis='x',
               labelsize=20,
               length=10,
               width=3,
               rotation=30)

ax.tick_params(axis='x',
               which='minor',
               length=5,
               width=2)

ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks,
              minor=True)

ax.tick_params(axis='y',
               labelsize=20,
               length=10,
               width=3)

ax.tick_params(axis='y',
               which='minor',
               length=5,
               width=2)
```

## 3.ax.set_xticklabels

### 3-1.Usage

```python
xticks = [i for i in range(10)]
xtick_labels = ['Class ' + str(i) for i in xticks]
yticks = [i for i in range(0,101,20)]
ytick_labels = [str(i) + '%' for i in yticks]

fig,ax = plt.subplots(figsize=(10,7))

ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels)
ax.tick_params(axis='x',
               labelsize=20,
               rotation=60)

ax.set_yticks(yticks)
ax.set_yticklabels(ytick_labels)
ax.tick_params(axis='y',
               labelsize=20)

fig.subplots_adjust(bottom=0.2,
                    left=0.15)
```

## 4.Practice

```python
# Practice 1
xticks = [i for i in range(7)]
xtick_labels = [' ','<vehicle>','<person>','<pet 1>','<pet 2>','<pet3>',' ']
yticks = [i/10 for i in range(11)]

fig,ax = plt.subplots(figsize=(10,7))

ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels,
                   ha='right')
ax.tick_params(axis='x',
               labelsize=20,
               rotation=30,
               colors='gray')

ax.set_yticks(yticks)
ax.tick_params(axis='y',
               labelsize=15,
               colors='gray')

fig.subplots_adjust(bottom=0.2,top=0.9,
                    left=0.1,right=0.9)
```

```python
# Practice 2
xticks = [i for i in range(7)]
xtick_labels = [' ','FC Agg 1 layer','FC Agg 2 layer','FC Agg 3 layer','FC Agg 4 layer','FC Agg 5 layer',' ']
yticks = np.arange(0.4,0.54,0.02)

fig,ax = plt.subplots(figsize=(7,7))

ax.set_xticks(xticks)
ax.set_xticklabels(xtick_labels,
                   ha='right')
ax.tick_params(axis='x',
               labelsize=20,
               rotation=30,
               colors='gray')

ax.set_yticks(yticks)
ax.set_ylim([0.4,0.52])
ax.tick_params(axis='y',
               labelsize=15,
               colors='gray')
ax.set_ylabel('mAP',
              fontsize=20,
              color='gray')
```

```python
# Practice 3
fig,ax = plt.subplots(figsize=(7,7))

ax.set_xscale('log')
ax.set_yscale('log')


major_xticks = [10**i for i in range(5)]
minor_xticks = [10**(i/10) for i in range(40)]
major_yticks = [10**i for i in range(-10,1,5)]
minor_yticks = [10**i for i in range(-10,4)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks,
              minor=True)
ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks,
              minor=True)


ax.tick_params(labelsize=20,
               length=10,
               width=3,
               direction='in')

ax.tick_params(which='minor',
               labelsize=0,
               length=5,
               width=1,
               direction='in')
```
