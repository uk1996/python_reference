```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1.ax grid

### 1-1.Axis Grid

```python
fig, axes = plt.subplots(1,3,figsize=(21,7))

axes[0].grid()
axes[1].grid(axis='x')
axes[2].grid(axis='y')
```

### 1-2.linwidth Argumnet


```python
fig, axes = plt.subplots(1,3,figsize=(21,7))

axes[0].grid(axis='x')
axes[1].grid(axis='x',
             linewidth = 2)
axes[2].grid(axis='x',
             linewidth = 5)
```

### 1-3.Line Styles in Matplotlib

```python
fig, axes = plt.subplots(1,3,figsize=(21,7))

axes[0].grid(axis='x',
             linewidth=3,
             linestyle = ':')
axes[1].grid(axis='x',
             linewidth=3,
             linestyle='-.')
axes[2].grid(axis='x',
             linewidth = 3,
             linestyle = '--')
```

### 1-4.Grid Color


```python
fig, axes = plt.subplots(1,3,figsize=(21,7))

axes[0].grid(axis='x',
             linewidth=2,
            color='r')
axes[1].grid(axis='x',
             linewidth=2,
             color='b')
axes[2].grid(axis='x',
             linewidth = 2,
             color='g')
```

## 2.Ticks and Grid

### 2-1.Major and Minor Grid

```python
fig,ax = plt.subplots(figsize=(7,7))

major_xticks = [i for i in range(0,101,20)]
minor_xticks = [i for i in range(0,101,5)]

ax.set_xticks(major_xticks)
ax.set_xticks(minor_xticks,minor = True)

ax.tick_params(labelsize=15)

ax.grid(axis='x',which = 'major',
        linewidth=2)
ax.grid(axis='x',which='minor',
        linestyle=':')
```

### 2-2.x,y Grid

```python
fig,ax = plt.subplots(figsize=(7,7))

major_yticks = [i for i in range(0,101,20)]
minor_yticks = [i for i in range(0,101,5)]
ytick_labels = [str(i) + '%' for i in major_yticks]

ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks,minor = True)

ax.set_yticklabels(ytick_labels)

ax.tick_params(labelsize=15)

ax.grid(axis='y',
        linewidth=2)
ax.grid(axis='y',which='minor',
        linestyle=':')
```

## 3.Practice

```python
# Practice 1

fig,ax = plt.subplots(figsize=(7,7))

ax.set_xscale('log')
ax.set_yscale('log')

ax.set_xlim([10**0,10**4])
ax.set_ylim([10**-10,10**3])


major_xticks = [10**i for i in range(5)]
major_yticks = [10**i for i in range(-10,1,5)]
minor_yticks = [10**i for i in range(-10,4)] 

ax.set_xticks(major_xticks)
ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks,
              minor = True)



ax.tick_params(which='major',
               labelsize=15,
               length=1,
               direction = 'in')
ax.tick_params(which='minor',
               length=5,
               labelsize = 0,
               direction='in')

ax.grid(which='major',
        linewidth=1.5,
        color='silver')
ax.grid(which = 'minor',
        linestyle=':',
        color='silver')
```

```python
# Practice 2

m_exp, M_exp = 0,4
n_inter_yticks = 4

n_major_yticks = M_exp - m_exp + 1
n_minor_yticks = (n_inter_yticks + 1)*(n_major_yticks - 1) + 1

major_xticks = np.logspace(0,4,5)
major_yticks = np.logspace(m_exp,M_exp,n_major_yticks)
minor_yticks = np.logspace(m_exp,M_exp,n_minor_yticks)

fig,ax = plt.subplots(figsize=(7,7))

ax.set_xscale('log')
ax.set_yscale('log')
ax.set_ylim([10**m_exp,10**M_exp])

ax.set_xticks(major_xticks)
ax.set_yticks(major_yticks)
ax.set_yticks(minor_yticks,
              minor=True)

ax.tick_params(axis='y',which='major',
               length=10)
ax.tick_params(axis='y',which='minor',
               length=5)

ax.grid(axis='y',which='major',
        linewidth = 2)
ax.grid(axis='y',which='minor',
        linestyle = ':')
```
