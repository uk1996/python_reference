```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1.Styles in Matplotlib

```python
plt.style.use('seaborn')

fig,axes = plt.subplots(2,1,figsize=(7,7))

axes[0].plot([1,2,3])
axes[1].scatter(np.random.normal(0,1,10,),
                np.random.normal(0,1,10,))
```

## 2.rcParams in Matplotlib
defalut 값을 설정

```python
plt.style.use('default')

plt.rcParams['figure.figsize'] = (8,4)
plt.rcParams['axes.titlesize'] = 30
plt.rcParams['axes.labelcolor'] = 'darkblue'
plt.rcParams['axes.labelsize'] = 20
plt.rcParams['axes.spines.right'] = False
plt.rcParams['axes.spines.top'] = False
plt.rcParams['grid.linestyle'] = ':'

fig, axes = plt.subplots(1,2)

axes[0].set_title('Title1')
axes[0].set_xlabel('X label1')
axes[0].set_ylabel('Y label1')
axes[0].grid()

axes[1].set_title('Title2')
axes[1].set_xlabel('X label2')
axes[1].set_ylabel('Y label2')
axes[1].grid()

fig.tight_layout()
```
