```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1.Spines



### 1-1.Spine Dictionary

```python
fig,ax = plt.subplots(figsize = (7,7))

print('Ax spines type:\n',type(ax.spines),'\n')
print('Ax spines:\n',ax.spines,'\n')
print('Ax spines keys:\n',ax.spines.keys(),'\n')
print('Ax spines values:\n',ax.spines.values(),'\n')
print('Ax spines items:\n',ax.spines.items(),'\n')
```

### 1-2.Get Spine Object

```python
fig, ax = plt.subplots(figsize=(7,7))

for spine_loc, spine in ax.spines.items():
    print(spine_loc,spine,sep = ' - ')

# 특정 spine에만 접근할때
for spine_loc,spine in ax.spines.items():
    if spine_loc in ['left','right']:
        pass   # if문안에 코드작성
    if spine_loc in ['bottom','top']:
        pass
```

## 2.spine.set_visible


### 2-1.Usage

```python
fig, ax = plt.subplots(figsize=(7,7))

for spine_loc,spine in ax.spines.items():
    if spine_loc in ['right','top']:
        spine.set_visible(False)
```

### 2-2.with Tick and Ticklabels

```python
fig,axes  = plt.subplots(1,2,figsize=(14,7))

for spine_loc,spine in axes[0].spines.items():
    if spine_loc in ['left','bottom']:
        spine.set_visible(False)

# ----------------------------------------------- 

for spine_loc,spine in axes[1].spines.items():
    if spine_loc in ['left','bottom']:
        spine.set_visible(False)

axes[1].tick_params(left=False,labelleft=False,
                    bottom=False,labelbottom=False)  
```

### 2-3.Upper Right Axis

```python
fig,ax = plt.subplots(figsize=(7,7))

for spine_loc,spine in ax.spines.items():
    if spine_loc in ['left','bottom']:
        spine.set_visible(False)

ax.tick_params(left=False,labelleft=False,
               bottom=False,labelbottom=False,
               right=True,labelright=True,
               top=True,labeltop=True)
```

### 2-4.Entire Spines

```python
np.random.seed(0)
noise_img = np.random.normal(0,1,(100,100))

fig, ax = plt.subplots(figsize=(7,7))

for spine_loc,spine in ax.spines.items():
    spine.set_visible(False)

ax.tick_params(left=False,labelleft=False,
               bottom=False,labelbottom=False)

ax.imshow(noise_img)
```

## 3.spine.set_position(Basic Usage)


### 3-1.Central Location

```python
fig,axes = plt.subplots(1,2,figsize=(14,7))

for spine_loc,spine in axes[0].spines.items():
    if spine_loc in ['top','bottom','right']:
        spine.set_visible(False)

axes[0].tick_params(bottom=False,labelbottom=False)

axes[0].spines['left'].set_position('center')

# ---------------------------------------------------
for spine_loc,spine in axes[1].spines.items():
    if spine_loc in ['top','right']:
        spine.set_visible(False)
    
    if spine_loc in ['left','bottom']:
        spine.set_position('center')
```

### 3-2.Variations

```python
fig,axes = plt.subplots(2,2,figsize=(7,7))

for spine_loc,spine in axes[0,0].spines.items():
    if spine_loc in ['top','right']:
        spine.set_visible(False)
    
    if spine_loc in ['left']:
        spine.set_position('center')

# -----------------------------------------------
for spine_loc,spine in axes[0,1].spines.items():
    if spine_loc in ['top','right']:
        spine.set_visible(False)
    
    if spine_loc in ['bottom']:
        spine.set_position('center')

# -----------------------------------------------
for spine_loc,spine in axes[1,0].spines.items():
    if spine_loc in ['bottom','right']:
        spine.set_visible(False)
    
    if spine_loc in ['left']:
        spine.set_position('center')

axes[1,0].tick_params(bottom=False,labelbottom=False,
                      top=True,labeltop=True)

# -----------------------------------------------
for spine_loc,spine in axes[1,1].spines.items():
    if spine_loc in ['top','left']:
        spine.set_visible(False)
    
    if spine_loc in ['bottom']:
        spine.set_position('center')
    
axes[1,1].tick_params(left=False,labelleft=False,
                      right=True,labelright=True)

fig.tight_layout()
```

## 4.spine.set_position(Tuple Argument)

### 4-1.Central Location with Tuple Argument

```python
fig, axes = plt.subplots(1,3,figsize=(15,5),facecolor='linen')

for i in range(3):
    for spine_loc,spine in axes[i].spines.items():
        if spine_loc in ['top','right','bottom']:
            spine.set_visible(False)

axes[0].spines['left'].set_position(('axes',0.5))
axes[1].spines['left'].set_position(('axes',0.2))
axes[2].spines['left'].set_position(('axes',0.8))        

fig.tight_layout()
```

```python
fig, axes = plt.subplots(1,2,figsize=(14,7))

np.random.seed(0)

t_min,t_max = -3,10
n_data = 100
t = np.linspace(t_min,t_max,n_data)
noise_data = np.random.normal(0,1,size=(n_data,)) - 2
axes[0].plot(t, noise_data)
axes[0].tick_params(labelsize=15)

x_data = np.random.normal(2, 3, (100,))
y_data = np.random.normal(2, 3, (100,))
axes[1].scatter(x_data,y_data)
axes[1].tick_params(labelsize=15)

fig.tight_layout()

for i in range(2):
    for spine_loc,spine in axes[i].spines.items():
        if spine_loc in ['right','top']:
            spine.set_visible(False)
    
        if spine_loc in ['bottom','left']:
            spine.set_position(('data',0))
```

### 4-2.Practice

```python
fig, ax = plt.subplots(figsize=(7,7))

x_data = np.random.normal(5, 3, (100,))
y_data = np.random.normal(2, 3, (100,))

ax.scatter(x_data,y_data)
ax.tick_params(labelsize=15)

fig.tight_layout()

for spine_loc,spine in ax.spines.items():
    if spine_loc in ['top','right']:
        spine.set_visible(False)
    
    if spine_loc in ['left']:
        spine.set_position(('data',5))
        spine.set_color('red')
        spine.set_alpha(0.5)
    
    if spine_loc in ['bottom']:
        spine.set_position(('data',2))
        spine.set_color('red')
        spine.set_alpha(0.5)
```
