```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1.Title and Label

### 1-1.fig.suptitle and ax.set_title

```python
fig,axes =plt.subplots(2,2,figsize=(7,7))

fig.suptitle('Title of a Figure',            # title
             fontsize=30,                  # 글자 크기 
             alpha = 0.5,                   # 투명도
             fontfamily='monospace')       # 글씨체

ax_title_list = ['title' + str(i) for i in range(4)]
for ax_idx,ax in enumerate(axes.flat):
    ax.set_title(ax_title_list[ax_idx],
                 fontsize=20,
                 fontfamily='serif')

fig.subplots_adjust(bottom=0.05, top=0.8, hspace=0.3)
```

### 1-2.ax.set_xlabel and ax.set_ylabel

```python
fig,ax = plt.subplots(figsize=(7,7))

ax.set_xlabel('X label',
              fontsize = 20,
              color = 'darkblue',
              alpha = 0.7)

ax.set_ylabel('Y label',
              fontsize = 20,
              color = 'teal',
              alpha = 0.7)
```

## 2.Text Alignment(ax.text)
- horizontal alignment or ha(center,left,right)
- vertical alignment or va(center,bottom,top)

```python
fig,ax = plt.subplots(figsize=(7,7))

ax.set_xlim([-1,1])
ax.set_ylim([-1,1])

ax.grid()
ax.tick_params(axis='both',
               labelsize = 15)

ax.text(x=0,y=0,
        va='center',ha='left',
         s='Hello1',
         fontsize=30)

ax.text(x=0,y=-0.25,
        va='center',ha='center',
         s='Hello2',
         fontsize=30)

ax.text(x=0.5,y=0.5,
        va='top',ha='right',
         s='Hello3',
         fontsize=30)

ax.text(x=-0.75,y=0,
        va='bottom',ha='center',
         s='Hello4 ',
         fontsize=30)

ax.text(x=-0.5,y=0.5,
        va='bottom',ha='right',
         s='Hello5',
         fontsize=30)
```

## 4.Text Properties

### 4-1.fontsize

```python
fig,ax = plt.subplots(figsize=(7,7))

fontsize_list = [20,30,40]

ax.set_xlim([-1,1])
ax.set_ylim([0,len(fontsize_list)+1])

for font_idx,fontsize in enumerate(fontsize_list):
    ax.text(x=0,y=font_idx+1,
            ha = 'center',
            s = 'fontsize = ' + str(fontsize),
            fontsize = fontsize)
```

### 4-2.fontfamily(글씨체)

```python
fig,ax = plt.subplots(figsize=(7,7))

fontfamily_list = ['serif','sans-serif','cursive','fantasy','monospace']

ax.set_xlim([-1,1])
ax.set_ylim([0,len(fontfamily_list)+1])

for font_idx,fontfamily in enumerate(fontfamily_list):
    ax.text(x=0,y=font_idx+1,
            ha='center',
            s=fontfamily,
            fontsize=30,
            fontfamily=fontfamily)
```

### 4-3.fontweight(굵기)

```python
fig,ax = plt.subplots(figsize=(7,7))

fontweight_list = ['normal','bold']

ax.set_xlim([-1,1])
ax.set_ylim([0,len(fontweight_list)+1])

for font_idx,fontweight in enumerate(fontweight_list):
    ax.text(x=0,y=font_idx+1,
            ha='center',
            s='fontweight = '+str(fontweight),
            fontsize=30,
            fontweight=fontweight)
```

### 4-4.alpha(투명도)

```python
fig,ax = plt.subplots(figsize=(7,7))

alpha_list = [0.1,0.3,0.5,0.7,0.9,1]

ax.set_xlim([-1,1])
ax.set_ylim([0,len(alpha_list)+1])

for a_idx,alpha in enumerate(alpha_list):
    ax.text(x=0,y=a_idx+1,
            ha='center',
            s='alpha = '+ str(alpha),
            fontsize=30,
            alpha=alpha)
```

### 4-5.color

```python
fig,ax = plt.subplots(figsize=(7,7))

color_list = ['red','darkred','chocolate','orange','olive','palegreen','skyblue','navy','indigo']

ax.set_xlim([-1,1])
ax.set_ylim([0,len(color_list)+1])

for color_idx,color in enumerate(color_list):
    ax.text(x=0,y=color_idx+1,
            ha='center',
            s=color,
            fontsize=30,
            color=color)
```

### 4-6.bbox

```python
fig,ax = plt.subplots(figsize=(7,7))

bbox_list = [{'boxstyle':'DArrow','color':'red','alpha':0.5},
             {'boxstyle':'Round','color':'orange','alpha':0.5},
             {'boxstyle':'Sawtooth','color':'teal','alpha':0.5},
             {'boxstyle':'DArrow','color':'navy','alpha':0.5}]

ax.set_xlim([-1,1])
ax.set_ylim([0,len(bbox_list)+1])

for b_idx,bbox in enumerate(bbox_list):
    ax.text(x=0,y=b_idx+1,
            ha='center',
            s=bbox,
            fontsize=15,
            bbox=bbox)
```

### 4-7.rotation

```python
fig,ax = plt.subplots(figsize=(7,7))

rotation_list = [0,15,30,45,60,75,90]

ax.set_xlim([0,len(rotation_list)+3])
ax.set_ylim([0,len(rotation_list)+3])

for r_idx,rotation in enumerate(rotation_list):
    ax.text(len(rotation_list)+1-r_idx,r_idx+1,
            ha='center',
            s='rotatoion = ' + str(rotation),
            rotation=rotation)
```

## 5.Font Dictonary

### 5-1.Usage

```python
np.random.seed(0)
fig,ax = plt.subplots(figsize=(7,7))

data = np.random.normal(0,1,(10,))
M_idx,M_val = np.argmax(data),np.round(np.max(data),2)
m_idx,m_val = np.argmin(data),np.round(np.min(data),2)

M_string = 'Max: ({},{})'.format(str(M_idx),str(M_val))
m_string = 'mix: ({},{})'.format(str(m_idx),str(m_val))

# --- font dictionary
title_font_dict = {'fontsize':30,'fontfamily':'serif',
                   'color':'darkblue','alpha':0.8}
xylabel_font_dict = {'fontsize':20,'fontfamily':'monospace',
                   'color':'darkblue','alpha':0.6}
M_font_dict = {'fontsize':20,'color':'red'}
m_font_dict = {'fontsize':20,'color':'blue'}
# ---

ax.plot(data)
ax.set_title('Random Data',
             fontdict = title_font_dict)
ax.set_xlabel('Data Index',
              fontdict=xylabel_font_dict)
ax.set_ylabel('Data Value',
              fontdict=xylabel_font_dict)
ax.text(x=M_idx+0.5,y=M_val-0.1,
        s=M_string,
        fontdict=M_font_dict)
ax.text(x=m_idx+0.5,y=m_val+0.1,
        s=m_string,
        fontdict=m_font_dict)
```

```python
# practice 2
fig = plt.figure(figsize=(14.4,6))

title_list = ['Blurry image','Ours','Mohan et al.']

axes = np.empty(shape=(0,4))

for g_idx in range(3):
    axes_g = np.empty(shape=(0,))
    main = plt.subplot2grid((5,12),(0,g_idx*4),
                            rowspan=3,colspan=4,
                            fig=fig)
    axes_g = np.append(axes_g,main)
    
    sub = plt.subplot2grid((5,12),(3,g_idx*4),
                           rowspan=2,colspan=3,
                           fig=fig)
    axes_g = np.append(axes_g,sub)

    for r_idx in range(3,3+2):
        ax = plt.subplot2grid((5,12),(r_idx,3+g_idx*4))
        axes_g = np.append(axes_g,ax)
    
    for ax in axes_g:
        ax.get_xaxis().set_visible(False)
        ax.get_yaxis().set_visible(False)
    
    axes = np.vstack((axes,axes_g))
    axes_g[0].set_title(title_list[g_idx],
                        fontsize=30,
                        y = -0.77,
                        va='top')

fig.subplots_adjust(bottom=0.1,top=0.95,
                   left=0.05,right=0.95,
                   wspace=0.05,hspace=0.05)
```
