```python
import matplotlib.pyplot as plt
import numpy as np
```

## 1.Named Colors

### 1-1.Base Colors

```python
color_list = ['b','g','r','c','m','y']

fig, ax = plt.subplots(figsize=(7,7))
ax.set_xlim([-1,1])
ax.set_ylim([-1,len(color_list)])

for c_idx,c in enumerate(color_list):
    ax.text(x = 0,y = c_idx,
            s = 'color = '+ c,
            fontsize=20,
            ha='center',
            color = c)
```

### 1-2.tab10 Colors

```python
color_list = ['tab:blue', 'tab:orange',
              'tab:green','tab:red',
              'tab:purple','tab:brown',
              'tab:pink','tab:gray',
              'tab:olive','tab:cyan']

fig, ax = plt.subplots(figsize=(7,7))
ax.set_xlim([-1,1])
ax.set_ylim([-1,len(color_list)])

for c_idx,c in enumerate(color_list):
    ax.text(x=0,y=c_idx,
            s='color = ' + c,
            fontsize=20,
            ha= 'center',
            color=c)
```

### 1-3.Practice

```python
title_font_dict = {'fontsize':20,
                   'color':'darkblue'}
label_font_dict = {'fontsize':20,
                   'color':'darkblue',
                   'alpha':0.8}

fig = plt.figure(figsize=(7,7),facecolor='linen')
ax = fig.add_subplot(facecolor='linen')

ax.set_title('Fig Title',
             fontdict = title_font_dict)
ax.set_xlabel('Data Index',
              fontdict = label_font_dict)
ax.set_ylabel('Data Values',
              fontdict = label_font_dict)

ax.tick_params(labelsize=15,
               colors='darkblue')

for spine_loc, spine in ax.spines.items():
    if spine_loc in ['top','right']:
        spine.set_visible(False)
    if spine_loc in ['left','bottom']:
        spine.set_color('darkblue')
    if spine_loc in ['bottom']:
        spine.set_position(('data',0))

test_np= np.random.normal(0,1,(10,))
ax.plot(test_np)

fig.tight_layout()
```

### 2-2.Byte to Float

```python
# --- 0~255 -> 0~1
def convert255to1(*RGB255):
    color = [c/255 for c in RGB255]
    return color
# ---

fig,ax = plt.subplots(figsize=(7,7))
ax.set_title('Fig Title',
             fontsize=30,
             color = convert255to1(200,100,150))
```

### 2-3.Practical Tip

```python
def convert255to1(*RGB255):
    color = [c/255 for c in RGB255]
    return color

ppt_color = convert255to1(128,85,128)
label_font_dict = {'color':'w',
                   'fontsize':20}

fig = plt.figure(figsize=(14,7),facecolor=ppt_color)

ax = fig.add_subplot(facecolor=ppt_color)

ax.plot(np.random.normal(0,1,(100,)),
        color='w')

ax.set_xlabel('Data Index',
              fontdict = label_font_dict)
ax.set_ylabel('Random Value',
              fontdict = label_font_dict)
ax.tick_params(labelsize=20,
               colors='w')

for spine_loc,spine in ax.spines.items():
    spine.set_color('w')

fig.subplots_adjust(bottom=0.1,top=0.9,
                    left=0.1,right=0.9)
```

## 3.Colormaps in Matplotlib

###  3-1.Discrete Colormaps

```python
import matplotlib.cm as cm

cmap = cm.get_cmap('tab10')
n_color = len(cmap.colors)
print('tab10 cmap:')
for c_idx in range(n_color):
    print(cmap(c_idx))
print('\n')

fig, ax = plt.subplots(figsize=(15,10))
ax.set_xlim([-1,1])
ax.set_ylim([-1,n_color])

for c_idx in range(n_color):
    color = cmap(c_idx)
    c_string_list = [c for c in str(color)]
    c_string = ''.join(c_string_list)

    ax.text(x=0,y=c_idx,
            s = 'color = ' + c_string,
            fontsize=15,
            ha='center',
            color = color)
```

### 3-2.Continuous Colormaps

```python
n_color = 10
cmap = cm.get_cmap('rainbow',lut=n_color) # lut Argument

print('rainbow cmap:')
for c_idx in range(n_color):
    print(cmap(c_idx))
print('\n')

fig, ax = plt.subplots(figsize=(15,10))
ax.set_xlim([-1,1])
ax.set_ylim([-1,n_color])

for c_idx in range(n_color):
    color = cmap(c_idx)
    c_string_list = [c for c in str(color)]
    c_string = ''.join(c_string_list)

    ax.text(x=0,y=c_idx,
            s = 'color = ' + c_string,
            fontsize=15,
            ha='center',
            color = color)
```

```python
# Auto Colormaps
harvest = np.array([[0.8,2.4,2.5],
                   [2.4,0.0,4.0],
                   [1.1,2.4,0.8]])

fig,axes = plt.subplots(2,2,figsize=(7,7))
im1 = axes[0,0].imshow(harvest)
im2 = axes[0,1].imshow(harvest,
                       cmap='Reds')
im3 = axes[1,0].imshow(harvest,
                       cmap='hot')
im4 = axes[1,1].imshow(harvest,
                       cmap='rainbow')
```

## 4.Colorbar

### 4-1.Setting Ticks

```python
x = np.linspace(-10,10,100)
y = x

X,Y = np.meshgrid(x, y)
Z = np.power(X,2) + np.power(Y,2)

levels = np.linspace(np.min(Z),np.max(Z),30)
cmap = cm.get_cmap('Reds_r',lut = len(levels))

fig, ax = plt.subplots(figsize=(8,7))
cf = ax.contourf(X,Y,Z,
                 levels=levels,
                 cmap=cmap)

cbar = fig.colorbar(cf) # color bar

cbar_ticks = cbar.get_ticks()
cbar_ticks = np.linspace(cbar_ticks[0],cbar_ticks[-1],3)

cbar.set_ticks(cbar_ticks)

cbar.set_ticklabels(['low','medium','high'])
cbar.ax.tick_params(labelsize=20)
```

### 4-2.Tick Location

```python
x = np.linspace(-10,10,100)
y = x

X,Y = np.meshgrid(x, y)
Z1 = np.power(X,2) + np.power(Y,2)
Z2 = np.power(X,3) + np.power(Y,3)

levels1 = np.linspace(np.min(Z1),np.max(Z1),30)
cmap1 = cm.get_cmap('Reds_r',lut = len(levels1))

levels2 = np.linspace(np.min(Z2),np.max(Z2),30)
cmap2 = cm.get_cmap('Blues_r',lut = len(levels2))

fig,axes = plt.subplots(1,2,figsize=(18,7))

cf1 = axes[0].contourf(X,Y,Z1,
                       levels=levels1,
                       cmap=cmap1)
cf2 = axes[1].contourf(X,Y,Z2,
                       levels=levels2,
                       cmap=cmap2)

cbar1 = fig.colorbar(cf1,ax=axes[0])
cbar2 = fig.colorbar(cf2,ax=axes[1])
```
